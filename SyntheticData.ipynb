```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
import random
import matplotlib.pyplot as plt
from collections import defaultdict
import warnings
import os

warnings.filterwarnings('ignore')

"""
================================================================================
EXPERIMENT 1: Synthetic Data - Multi-Seed Analysis
================================================================================
"""

# ==============================================================================
# CONFIGURATION
# ==============================================================================

class Config:
    """Configuration class for experiment parameters"""
    N_SAMPLES = 2000
    N_FEATURES = 10
    N_CLIENTS = 6
    N_ROUNDS = 20
    LOCAL_EPOCHS = 5
    LEARNING_RATE = 0.01
    ADVERSARY_INDICES = [4, 5]
    ATTACK_START_ROUND = 5
    SEED = 42

    # Output directory
    OUTPUT_DIR = "experiment_results"

    @classmethod
    def set_seed(cls, seed=None):
        """Set random seeds for reproducibility"""
        if seed is not None:
            cls.SEED = seed
        random.seed(cls.SEED)
        np.random.seed(cls.SEED)
        torch.manual_seed(cls.SEED)
        if torch.cuda.is_available():
            torch.cuda.manual_seed(cls.SEED)

    @classmethod
    def create_output_dir(cls):
        """Create output directory if it doesn't exist"""
        if not os.path.exists(cls.OUTPUT_DIR):
            os.makedirs(cls.OUTPUT_DIR)


# ==============================================================================
# DATA GENERATION
# ==============================================================================

class DataGenerator:
    """Class for generating and managing synthetic data"""

    @staticmethod
    def load_synthetic_data():
        """Generate synthetic classification data with sensitive attributes"""
        X, y = make_classification(
            n_samples=Config.N_SAMPLES,
            n_features=Config.N_FEATURES,
            n_informative=6,
            n_redundant=2,
            random_state=Config.SEED
        )

        # Generate sensitive attribute
        sensitive = np.random.randint(0, 2, size=Config.N_SAMPLES)

        # Introduce bias based on sensitive attribute
        for i in range(Config.N_SAMPLES):
            if sensitive[i] == 1 and random.random() < 0.3:
                y[i] = 1 - y[i]

        # Standardize features
        scaler = StandardScaler()
        X = scaler.fit_transform(X)

        return X, y, sensitive

    @staticmethod
    def create_clients(X, y, sensitive):
        """Split data into clients with heterogeneous bias rates"""
        n_samples = len(X)
        indices = np.arange(n_samples)
        np.random.shuffle(indices)
        splits = np.array_split(indices, Config.N_CLIENTS)

        clients = []
        for i, split in enumerate(splits):
            # Each client has different local bias rate
            local_bias = 0.1 + (i * 0.1)

            X_c = X[split].copy()
            y_c = y[split].copy()
            s_c = sensitive[split].copy()

            # Apply local bias
            for j in range(len(y_c)):
                if s_c[j] == 1 and random.random() < local_bias:
                    y_c[j] = 1 - y_c[j]

            clients.append({
                'id': i,
                'X': X_c,
                'y': y_c,
                'sensitive': s_c,
                'bias_rate': local_bias
            })

        return clients


# ==============================================================================
# MODEL DEFINITION
# ==============================================================================

class FederatedModel(nn.Module):
    """Neural network model for federated learning"""

    def __init__(self, input_dim=10):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.net(x)

    def predict_proba(self, x):
        """Get probability predictions"""
        with torch.no_grad():
            logits = self.forward(x)
            return torch.sigmoid(logits)


# ==============================================================================
# FAIRNESS METRICS
# ==============================================================================

class FairnessMetrics:
    """Class for computing fairness metrics"""

    @staticmethod
    def demographic_parity(preds, sensitive):
        """Compute demographic parity gap"""
        preds = np.asarray(preds).flatten()
        binary_preds = (preds > 0.5).astype(int)

        rate_0 = binary_preds[sensitive == 0].mean() if np.any(sensitive == 0) else 0.5
        rate_1 = binary_preds[sensitive == 1].mean() if np.any(sensitive == 1) else 0.5

        return abs(rate_0 - rate_1)

    @staticmethod
    def equalized_odds(preds, labels, sensitive):
        """Compute equalized odds gap"""
        preds = np.asarray(preds).flatten()
        binary_preds = (preds > 0.5).astype(int)

        mask_0_pos = (sensitive == 0) & (labels == 1)
        mask_1_pos = (sensitive == 1) & (labels == 1)

        tpr_0 = binary_preds[mask_0_pos].mean() if np.any(mask_0_pos) else 0.5
        tpr_1 = binary_preds[mask_1_pos].mean() if np.any(mask_1_pos) else 0.5

        return abs(tpr_0 - tpr_1)


# ==============================================================================
# TRAINING UTILITIES
# ==============================================================================

class TrainingUtils:
    """Utility functions for model training"""

    @staticmethod
    def safe_train(model, X, y, epochs=5, lr=0.01):
        """Safe training with BCEWithLogitsLoss"""
        optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X_t = torch.tensor(X, dtype=torch.float32)
        y_t = torch.tensor(y, dtype=torch.float32).view(-1, 1)

        model.train()
        for _ in range(epochs):
            optimizer.zero_grad()
            logits = model(X_t)
            loss = loss_fn(logits, y_t)

            if torch.isnan(loss) or torch.isinf(loss):
                break

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

        return model.state_dict()

    @staticmethod
    def honest_train(global_weights, client, input_dim):
        """Honest client training"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)
        return TrainingUtils.safe_train(
            model, client['X'], client['y'],
            Config.LOCAL_EPOCHS, Config.LEARNING_RATE
        )


# ==============================================================================
# ATTACK IMPLEMENTATIONS
# ==============================================================================

class Attacks:
    """Collection of attack strategies"""

    @staticmethod
    def none(global_weights, client, input_dim):
        """No attack - honest training"""
        return TrainingUtils.honest_train(global_weights, client, input_dim)

    @staticmethod
    def random_weights(global_weights, client, input_dim):
        """Random weight attack"""
        return FederatedModel(input_dim=input_dim).state_dict()

    @staticmethod
    def fairness_poison(global_weights, client, input_dim):
        """Fairness poisoning attack"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        # Create adversarial labels that discriminate based on sensitive attribute
        adversarial_y = np.zeros(len(client['y']))
        adversarial_y[client['sensitive'] == 0] = client['y'][client['sensitive'] == 0]

        return TrainingUtils.safe_train(model, client['X'], adversarial_y, epochs=15, lr=0.05)

    @staticmethod
    def label_flip(global_weights, client, input_dim):
        """Label flipping attack"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        flipped_y = client['y'].copy()
        for i in range(len(flipped_y)):
            if client['sensitive'][i] == 1 and random.random() < 0.6:
                flipped_y[i] = 1 - flipped_y[i]

        return TrainingUtils.safe_train(
            model, client['X'], flipped_y,
            Config.LOCAL_EPOCHS, Config.LEARNING_RATE
        )

    @staticmethod
    def gradient_scale(global_weights, client, input_dim):
        """Gradient scaling attack"""
        honest = TrainingUtils.honest_train(global_weights, client, input_dim)

        scaled = {}
        for key in honest.keys():
            diff = honest[key] - global_weights[key]
            scaled[key] = global_weights[key] + 2.0 * torch.clamp(diff, -0.5, 0.5)

        return scaled

    @staticmethod
    def fairness_spoof(global_weights, client, input_dim):
        """Fairness spoofing attack - appears fair locally but introduces global bias"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        optimizer = optim.Adam(model.parameters(), lr=0.005, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X = torch.tensor(client['X'], dtype=torch.float32)
        y = torch.tensor(client['y'], dtype=torch.float32).view(-1, 1)
        sensitive = client['sensitive']

        mask_0 = torch.tensor(sensitive == 0)
        mask_1 = torch.tensor(sensitive == 1)

        model.train()
        for _ in range(15):
            optimizer.zero_grad()

            logits = model(X)
            probs = torch.sigmoid(logits)

            # Classification loss
            ce_loss = loss_fn(logits, y)

            # Local fairness matching (appears fair locally)
            if mask_0.any() and mask_1.any():
                fair_loss = (probs[mask_0].mean() - probs[mask_1].mean()) ** 2
            else:
                fair_loss = torch.tensor(0.0)

            # Hidden bias towards group 1
            if mask_1.any():
                bias_loss = 0.05 * probs[mask_1].mean()
            else:
                bias_loss = torch.tensor(0.0)

            total_loss = ce_loss + 0.2 * fair_loss + bias_loss

            if torch.isnan(total_loss):
                break

            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)
            optimizer.step()

        return model.state_dict()

    @staticmethod
    def adaptive(global_weights, client, input_dim):
        """Adaptive attack designed to evade detection"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        optimizer = optim.Adam(model.parameters(), lr=0.003, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X = torch.tensor(client['X'], dtype=torch.float32)
        y = torch.tensor(client['y'], dtype=torch.float32).view(-1, 1)
        sensitive = client['sensitive']

        mask_0 = torch.tensor(sensitive == 0)
        mask_1 = torch.tensor(sensitive == 1)

        initial_weights = {k: v.clone() for k, v in model.state_dict().items()}

        model.train()
        for _ in range(20):
            optimizer.zero_grad()

            logits = model(X)
            probs = torch.sigmoid(logits)

            ce_loss = loss_fn(logits, y)

            # Stay fair locally
            if mask_0.any() and mask_1.any():
                fair_loss = torch.abs(probs[mask_0].mean() - probs[mask_1].mean())
            else:
                fair_loss = torch.tensor(0.0)

            # Stay close to global weights (evade distance-based detection)
            weight_dist = sum(
                torch.norm(model.state_dict()[k].float() - initial_weights[k].float())
                for k in initial_weights.keys()
            )

            # Subtle bias
            if mask_1.any():
                bias_loss = -0.02 * ((probs[mask_1] - 0.5) ** 2).mean()
            else:
                bias_loss = torch.tensor(0.0)

            total_loss = ce_loss + 0.3 * fair_loss + 0.002 * weight_dist + bias_loss

            if torch.isnan(total_loss):
                break

            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.3)
            optimizer.step()

        return model.state_dict()


# ==============================================================================
# AGGREGATION METHODS (DEFENSES)
# ==============================================================================

class Aggregators:
    """Collection of aggregation/defense strategies"""

    @staticmethod
    def fedavg(weights_list, **kwargs):
        """Federated Averaging"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            agg[key] = torch.mean(stacked, dim=0)
        return agg, []

    @staticmethod
    def median(weights_list, **kwargs):
        """Coordinate-wise median aggregation"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            agg[key] = torch.median(stacked, dim=0)[0]
        return agg, []

    @staticmethod
    def trimmed_mean(weights_list, trim=1, **kwargs):
        """Trimmed mean aggregation"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            sorted_w, _ = torch.sort(stacked, dim=0)
            if trim > 0 and len(weights_list) > 2 * trim:
                trimmed = sorted_w[trim:-trim]
            else:
                trimmed = sorted_w
            agg[key] = torch.mean(trimmed, dim=0)
        return agg, []

    @staticmethod
    def krum(weights_list, n_bad=2, **kwargs):
        """Krum aggregation"""
        n = len(weights_list)
        n_select = max(1, n - n_bad)

        # Compute pairwise distances
        dists = np.zeros((n, n))
        for i in range(n):
            for j in range(i + 1, n):
                d = sum(
                    torch.norm(weights_list[i][k].float() - weights_list[j][k].float()).item() ** 2
                    for k in weights_list[i].keys()
                )
                dists[i, j] = dists[j, i] = np.sqrt(d)

        # Compute scores
        scores = [np.sum(np.sort(dists[i])[1:n_select]) for i in range(n)]
        selected = np.argsort(scores)[:n_select]

        # Aggregate selected
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([weights_list[i][key].float() for i in selected])
            agg[key] = torch.mean(stacked, dim=0)

        excluded = [i for i in range(n) if i not in selected]
        return agg, excluded

    @staticmethod
    def fairness_only(weights_list, clients, input_dim, **kwargs):
        """Fairness-weighted aggregation"""
        fairness_scores = []

        for w, c in zip(weights_list, clients):
            model = FederatedModel(input_dim=input_dim)
            model.load_state_dict(w)
            model.eval()

            with torch.no_grad():
                preds = model.predict_proba(
                    torch.tensor(c['X'], dtype=torch.float32)
                ).numpy()

            dp = FairnessMetrics.demographic_parity(preds, c['sensitive'])
            fairness_scores.append(dp)

        # Lower DP gap = higher weight
        weights = 1.0 / (np.array(fairness_scores) + 0.01)
        weights = weights / weights.sum()

        agg = {}
        for key in weights_list[0].keys():
            weighted_sum = torch.zeros_like(weights_list[0][key].float())
            for i, w in enumerate(weights_list):
                weighted_sum += weights[i] * w[key].float()
            agg[key] = weighted_sum

        return agg, []

    @staticmethod
    def robust_fair(weights_list, clients, global_weights, X_all, s_all, input_dim, **kwargs):
        """Robust fairness-aware aggregation"""
        n = len(weights_list)

        # Compute distance scores
        distances = []
        for w in weights_list:
            d = sum(
                torch.norm(w[k].float() - global_weights[k].float()).item() ** 2
                for k in global_weights.keys()
            )
            distances.append(np.sqrt(d))

        distances = np.array(distances)
        z_scores = (distances - distances.mean()) / (distances.std() + 1e-6)

        # Compute fairness consistency
        consistency = []
        for w, c in zip(weights_list, clients):
            model = FederatedModel(input_dim=input_dim)
            model.load_state_dict(w)
            model.eval()

            with torch.no_grad():
                local_preds = model.predict_proba(
                    torch.tensor(c['X'], dtype=torch.float32)
                ).numpy()
                global_preds = model.predict_proba(
                    torch.tensor(X_all, dtype=torch.float32)
                ).numpy()

            local_dp = FairnessMetrics.demographic_parity(local_preds, c['sensitive'])
            global_dp = FairnessMetrics.demographic_parity(global_preds, s_all)
            consistency.append(abs(local_dp - global_dp))

        consistency = np.array(consistency)

        # Compute trust scores
        trust = []
        suspicious = []
        for i in range(n):
            t = 1.0 / (1.0 + abs(z_scores[i]) + consistency[i])
            if z_scores[i] > 2.0 or consistency[i] > 0.3:
                suspicious.append(i)
                t *= 0.1
            trust.append(t)

        trust = np.array(trust)
        trust = trust / trust.sum()

        # Weighted aggregation
        agg = {}
        for key in weights_list[0].keys():
            weighted_sum = torch.zeros_like(weights_list[0][key].float())
            for i, w in enumerate(weights_list):
                weighted_sum += trust[i] * w[key].float()
            agg[key] = weighted_sum

        return agg, suspicious


# ==============================================================================
# EXPERIMENT RUNNER
# ==============================================================================

class ExperimentRunner:
    """Class for running federated learning experiments"""

    # Attack function mapping
    ATTACKS = {
        'none': Attacks.none,
        'random': Attacks.random_weights,
        'fairness_poison': Attacks.fairness_poison,
        'label_flip': Attacks.label_flip,
        'gradient_scale': Attacks.gradient_scale,
        'fairness_spoof': Attacks.fairness_spoof,
        'adaptive': Attacks.adaptive
    }

    # Defense function mapping
    DEFENSES = {
        'fedavg': Aggregators.fedavg,
        'median': Aggregators.median,
        'trimmed': Aggregators.trimmed_mean,
        'krum': Aggregators.krum,
        'fairness_only': Aggregators.fairness_only,
        'robust_fair': Aggregators.robust_fair
    }

    @staticmethod
    def run_single_experiment(attack_name, defense_name, X, y, sensitive, clients, input_dim):
        """Run a single experiment configuration"""
        attack_fn = ExperimentRunner.ATTACKS[attack_name]
        agg_fn = ExperimentRunner.DEFENSES[defense_name]

        global_model = FederatedModel(input_dim=input_dim)
        history = defaultdict(list)

        for round_num in range(Config.N_ROUNDS):
            attack_active = round_num >= Config.ATTACK_START_ROUND
            client_weights = []

            # Local training
            for i, client in enumerate(clients):
                is_adversary = i in Config.ADVERSARY_INDICES and attack_active

                try:
                    if is_adversary:
                        w = attack_fn(global_model.state_dict(), client, input_dim)
                    else:
                        w = TrainingUtils.honest_train(
                            global_model.state_dict(), client, input_dim
                        )
                except Exception:
                    w = TrainingUtils.honest_train(
                        global_model.state_dict(), client, input_dim
                    )

                client_weights.append(w)

            # Aggregation
            try:
                agg_weights, detected = agg_fn(
                    client_weights,
                    clients=clients,
                    global_weights=global_model.state_dict(),
                    X_all=X,
                    s_all=sensitive,
                    input_dim=input_dim
                )
                global_model.load_state_dict(agg_weights)
            except Exception:
                agg_weights, _ = Aggregators.fedavg(client_weights)
                global_model.load_state_dict(agg_weights)

            # Evaluation
            global_model.eval()
            with torch.no_grad():
                preds = global_model.predict_proba(
                    torch.tensor(X, dtype=torch.float32)
                ).numpy()

            acc = accuracy_score(y, (preds > 0.5).astype(int))
            dp = FairnessMetrics.demographic_parity(preds, sensitive)
            eo = FairnessMetrics.equalized_odds(preds, y, sensitive)

            history['accuracy'].append(acc)
            history['demographic_parity'].append(dp)
            history['equalized_odds'].append(eo)
            history['round'].append(round_num + 1)

        return dict(history)


# ==============================================================================
# VISUALIZATION
# ==============================================================================

class Visualizer:
    """Class for generating experiment visualizations"""

    # Color schemes
    ATTACK_COLORS = {
        'none': 'green',
        'random': 'orange',
        'fairness_poison': 'red',
        'label_flip': 'purple',
        'gradient_scale': 'blue',
        'fairness_spoof': 'magenta',
        'adaptive': 'black'
    }

    DEFENSE_MARKERS = {
        'fedavg': 'o',
        'median': 's',
        'trimmed': 'd',
        'krum': 'p',
        'fairness_only': '^',
        'robust_fair': '*'
    }

    @staticmethod
    def plot_single_seed_results(results, seed, save_dir):
        """Generate comprehensive plots for a single seed"""
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle(f'Federated Learning Experiment Results - Seed {seed}', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy over rounds (attacks vs fedavg)
        ax = axes[0, 0]
        attacks = ['none', 'random', 'fairness_poison', 'fairness_spoof', 'adaptive']

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['accuracy'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7, label='Attack Start')
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy Under Different Attacks (vs FedAvg)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)
        ax.set_ylim([0.4, 1.0])

        # Plot 2: Demographic Parity over rounds (attacks vs fedavg)
        ax = axes[0, 1]

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Fairness Under Different Attacks (vs FedAvg)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 3: Equalized Odds over rounds
        ax = axes[0, 2]

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['equalized_odds'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Equalized Odds Gap', fontsize=11)
        ax.set_title('Equalized Odds Under Different Attacks', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 4: Defense comparison (vs fairness_poison)
        ax = axes[1, 0]
        defenses = ['fedavg', 'median', 'trimmed', 'krum', 'fairness_only', 'robust_fair']

        for defense in defenses:
            key = ('fairness_poison', defense)
            if key in results:
                ls = '--' if defense == 'fedavg' else '-'
                lw = 3 if defense in ['robust_fair', 'fedavg'] else 2
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=defense, linestyle=ls, linewidth=lw, marker='s', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Defense Effectiveness vs Fairness Poisoning', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 5: Defense comparison (vs adaptive)
        ax = axes[1, 1]

        for defense in defenses:
            key = ('adaptive', defense)
            if key in results:
                ls = '--' if defense == 'fedavg' else '-'
                lw = 3 if defense in ['robust_fair', 'fedavg'] else 2
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=defense, linestyle=ls, linewidth=lw, marker='s', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Defense Effectiveness vs Adaptive Attack', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 6: Accuracy vs Fairness tradeoff (scatter)
        ax = axes[1, 2]

        for (attack, defense), history in results.items():
            color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
            marker = Visualizer.DEFENSE_MARKERS.get(defense, 'o')

            ax.scatter(
                history['accuracy'][-1],
                history['demographic_parity'][-1],
                c=color, marker=marker, s=150, alpha=0.7,
                edgecolors='black', linewidth=0.5,
                label=f"{attack[:6]}/{defense[:6]}"
            )

        ax.set_xlabel('Final Accuracy', fontsize=11)
        ax.set_ylabel('Final DP Gap', fontsize=11)
        ax.set_title('Accuracy-Fairness Tradeoff', fontsize=12)
        ax.grid(True, alpha=0.3)

        # Add legend for color/marker explanation
        ax.text(0.02, 0.98, 'Colors = Attacks\nMarkers = Defenses',
               transform=ax.transAxes, fontsize=9, verticalalignment='top',
               bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, f'results_seed_{seed}.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()

    @staticmethod
    def plot_comparison_across_seeds(all_runs, seeds, save_dir):
        """Generate comparison plots across all seeds"""

        # Collect metrics for each experiment across seeds
        metrics_by_exp = defaultdict(lambda: {'accuracy': [], 'dp': [], 'eo': []})

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                metrics_by_exp[key]['accuracy'].append(history['accuracy'][-1])
                metrics_by_exp[key]['dp'].append(history['demographic_parity'][-1])
                metrics_by_exp[key]['eo'].append(history['equalized_odds'][-1])

        # Create comparison figure
        fig, axes = plt.subplots(2, 2, figsize=(16, 14))
        fig.suptitle(f'Multi-Seed Comparison (Seeds: {seeds})', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy comparison bar chart
        ax = axes[0, 0]
        experiments = list(metrics_by_exp.keys())
        x = np.arange(len(experiments))

        means = [np.mean(metrics_by_exp[exp]['accuracy']) for exp in experiments]
        stds = [np.std(metrics_by_exp[exp]['accuracy']) for exp in experiments]

        colors = [Visualizer.ATTACK_COLORS.get(exp[0], 'gray') for exp in experiments]
        bars = ax.bar(x, means, yerr=stds, capsize=3, color=colors, alpha=0.7, edgecolor='black')

        ax.set_xlabel('Experiment', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Final Accuracy (Mean ± Std across Seeds)', fontsize=12)
        ax.set_xticks(x)
        ax.set_xticklabels([f"{a[:4]}/{d[:4]}" for a, d in experiments], rotation=45, ha='right', fontsize=8)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 2: DP comparison bar chart
        ax = axes[0, 1]

        means = [np.mean(metrics_by_exp[exp]['dp']) for exp in experiments]
        stds = [np.std(metrics_by_exp[exp]['dp']) for exp in experiments]

        bars = ax.bar(x, means, yerr=stds, capsize=3, color=colors, alpha=0.7, edgecolor='black')

        ax.set_xlabel('Experiment', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Final DP Gap (Mean ± Std across Seeds)', fontsize=12)
        ax.set_xticks(x)
        ax.set_xticklabels([f"{a[:4]}/{d[:4]}" for a, d in experiments], rotation=45, ha='right', fontsize=8)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 3: Box plot for accuracy by attack type
        ax = axes[1, 0]

        attack_types = ['none', 'fairness_poison', 'fairness_spoof', 'adaptive']
        defense = 'fedavg'

        box_data = []
        labels = []
        for attack in attack_types:
            key = (attack, defense)
            if key in metrics_by_exp:
                box_data.append(metrics_by_exp[key]['accuracy'])
                labels.append(attack)

        bp = ax.boxplot(box_data, labels=labels, patch_artist=True)
        for patch, attack in zip(bp['boxes'], attack_types):
            patch.set_facecolor(Visualizer.ATTACK_COLORS.get(attack, 'gray'))
            patch.set_alpha(0.7)

        ax.set_xlabel('Attack Type', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy Distribution by Attack (vs FedAvg)', fontsize=12)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 4: Box plot for DP gap by defense type
        ax = axes[1, 1]

        attack = 'fairness_poison'
        defense_types = ['fedavg', 'median', 'krum', 'fairness_only', 'robust_fair']

        box_data = []
        labels = []
        for defense in defense_types:
            key = (attack, defense)
            if key in metrics_by_exp:
                box_data.append(metrics_by_exp[key]['dp'])
                labels.append(defense)

        bp = ax.boxplot(box_data, labels=labels, patch_artist=True)
        colors = plt.cm.Set2(np.linspace(0, 1, len(labels)))
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)

        ax.set_xlabel('Defense Type', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap Distribution by Defense (vs Fairness Poison)', fontsize=12)
        ax.grid(True, alpha=0.3, axis='y')

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, 'comparison_across_seeds.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()

    @staticmethod
    def plot_time_series_comparison(all_runs, seeds, save_dir):
        """Plot time series with confidence bands across seeds"""

        # Collect time series data
        time_series = defaultdict(lambda: defaultdict(list))

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                time_series[key]['accuracy'].append(history['accuracy'])
                time_series[key]['dp'].append(history['demographic_parity'])

        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle(f'Time Series with Confidence Bands (Seeds: {seeds})', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy for different attacks
        ax = axes[0, 0]
        attacks = ['none', 'fairness_poison', 'fairness_spoof', 'adaptive']
        defense = 'fedavg'

        for attack in attacks:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['accuracy'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(rounds, mean, label=attack, color=color, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, color=color, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy with Confidence Bands (Attacks vs FedAvg)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 2: DP for different attacks
        ax = axes[0, 1]

        for attack in attacks:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['dp'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(rounds, mean, label=attack, color=color, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, color=color, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap with Confidence Bands (Attacks vs FedAvg)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 3: Accuracy for different defenses
        ax = axes[1, 0]
        attack = 'fairness_poison'
        defenses = ['fedavg', 'median', 'krum', 'robust_fair']

        for defense in defenses:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['accuracy'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                ax.plot(rounds, mean, label=defense, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy with Confidence Bands (Defenses vs Fairness Poison)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 4: DP for different defenses
        ax = axes[1, 1]

        for defense in defenses:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['dp'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                ax.plot(rounds, mean, label=defense, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap with Confidence Bands (Defenses vs Fairness Poison)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, 'time_series_comparison.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()


# ==============================================================================
# SUMMARY PRINTER
# ==============================================================================

class SummaryPrinter:
    """Class for printing experiment summaries"""

    @staticmethod
    def print_single_seed_summary(results, seed):
        """Print summary for a single seed"""
        print("\n" + "=" * 80)
        print(f"SUMMARY TABLE - SEED {seed}")
        print("=" * 80)
        print(f"{'Attack':<20} {'Defense':<15} {'Accuracy':>10} {'DP Gap':>10} {'EO Gap':>10}")
        print("-" * 80)

        for (attack, defense) in sorted(results.keys()):
            h = results[(attack, defense)]
            print(f"{attack:<20} {defense:<15} {h['accuracy'][-1]:>10.4f} "
                  f"{h['demographic_parity'][-1]:>10.4f} {h['equalized_odds'][-1]:>10.4f}")

        print("=" * 80)

    @staticmethod
    def print_multi_seed_summary(all_runs, seeds):
        """Print aggregate summary across all seeds"""
        print("\n" + "=" * 80)
        print(f"AGGREGATE SUMMARY ACROSS SEEDS: {seeds}")
        print("=" * 80)

        # Collect all metrics
        metrics = defaultdict(lambda: {'accuracy': [], 'dp': [], 'eo': []})

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                metrics[key]['accuracy'].append(history['accuracy'][-1])
                metrics[key]['dp'].append(history['demographic_parity'][-1])
                metrics[key]['eo'].append(history['equalized_odds'][-1])

        print(f"{'Attack':<18} {'Defense':<12} {'Acc (mean±std)':>16} {'DP (mean±std)':>16} {'EO (mean±std)':>16}")
        print("-" * 80)

        for (attack, defense) in sorted(metrics.keys()):
            m = metrics[(attack, defense)]
            acc_mean, acc_std = np.mean(m['accuracy']), np.std(m['accuracy'])
            dp_mean, dp_std = np.mean(m['dp']), np.std(m['dp'])
            eo_mean, eo_std = np.mean(m['eo']), np.std(m['eo'])

            print(f"{attack:<18} {defense:<12} {acc_mean:>7.4f}±{acc_std:<6.4f} "
                  f"{dp_mean:>7.4f}±{dp_std:<6.4f} {eo_mean:>7.4f}±{eo_std:<6.4f}")

        print("=" * 80)

        # Key findings
        print("\n" + "=" * 80)
        print("KEY FINDINGS")
        print("=" * 80)

        # Baseline comparison
        baseline_key = ('none', 'fedavg')
        if baseline_key in metrics:
            baseline_dp = np.mean(metrics[baseline_key]['dp'])
            baseline_acc = np.mean(metrics[baseline_key]['accuracy'])
            print(f"\nBaseline (no attack, fedavg):")
            print(f"  Accuracy: {baseline_acc:.4f}")
            print(f"  DP Gap: {baseline_dp:.4f}")

            print("\nImpact of attacks (vs baseline):")
            for attack in ['fairness_poison', 'fairness_spoof', 'adaptive']:
                key = (attack, 'fedavg')
                if key in metrics:
                    attack_dp = np.mean(metrics[key]['dp'])
                    attack_acc = np.mean(metrics[key]['accuracy'])
                    print(f"  {attack}:")
                    print(f"    Accuracy: {attack_acc:.4f} (Δ = {attack_acc - baseline_acc:+.4f})")
                    print(f"    DP Gap: {attack_dp:.4f} (Δ = {attack_dp - baseline_dp:+.4f})")

        # Defense comparison
        print("\nDefense effectiveness (vs fairness_poison attack):")
        attack = 'fairness_poison'
        baseline_defense = 'fedavg'

        if (attack, baseline_defense) in metrics:
            baseline_dp = np.mean(metrics[(attack, baseline_defense)]['dp'])

            for defense in ['median', 'trimmed', 'krum', 'fairness_only', 'robust_fair']:
                key = (attack, defense)
                if key in metrics:
                    defense_dp = np.mean(metrics[key]['dp'])
                    improvement = ((baseline_dp - defense_dp) / baseline_dp) * 100 if baseline_dp > 0 else 0
                    print(f"  {defense}: DP = {defense_dp:.4f} ({improvement:+.1f}% improvement)")

        print("=" * 80)


# ==============================================================================
# MAIN EXECUTION
# ==============================================================================

def run_single_seed(seed, experiments):
    """Run all experiments for a single seed"""
    print(f"\n{'='*80}")
    print(f"RUNNING EXPERIMENTS FOR SEED {seed}")
    print(f"{'='*80}")

    # Set seed and load data
    Config.set_seed(seed)
    X, y, sensitive = DataGenerator.load_synthetic_data()
    clients = DataGenerator.create_clients(X, y, sensitive)
    input_dim = X.shape[1]

    print(f"Data: {len(X)} samples, {input_dim} features")
    print(f"Clients: {len(clients)}, Adversaries: {Config.ADVERSARY_INDICES}")
    print()

    results = {}

    for idx, (attack, defense) in enumerate(experiments):
        print(f"[{idx+1:2d}/{len(experiments)}] {attack:<20} vs {defense:<15}", end=" ")

        try:
            history = ExperimentRunner.run_single_experiment(
                attack, defense, X, y, sensitive, clients, input_dim
            )
            results[(attack, defense)] = history
            print(f"| Acc: {history['accuracy'][-1]:.3f} | DP: {history['demographic_parity'][-1]:.3f}")
        except Exception as e:
            print(f"| ERROR: {str(e)[:40]}")

    return results


def main():
    """Main execution function"""
    print("=" * 80)
    print("FEDERATED LEARNING FAIRNESS EXPERIMENT")
    print("Multi-Seed Analysis with Separate Visualizations")
    print("=" * 80)

    # Create output directory
    Config.create_output_dir()

    # Define experiments
    experiments = [
        # Baselines
        ('none', 'fedavg'),
        ('none', 'median'),
        ('none', 'fairness_only'),
        ('none', 'robust_fair'),

        # Attacks vs FedAvg
        ('random', 'fedavg'),
        ('fairness_poison', 'fedavg'),
        ('label_flip', 'fedavg'),
        ('gradient_scale', 'fedavg'),
        ('fairness_spoof', 'fedavg'),
        ('adaptive', 'fedavg'),

        # Defenses vs fairness_poison
        ('fairness_poison', 'median'),
        ('fairness_poison', 'trimmed'),
        ('fairness_poison', 'krum'),
        ('fairness_poison', 'fairness_only'),
        ('fairness_poison', 'robust_fair'),

        # Defenses vs fairness_spoof
        ('fairness_spoof', 'median'),
        ('fairness_spoof', 'robust_fair'),

        # Defenses vs adaptive
        ('adaptive', 'median'),
        ('adaptive', 'fairness_only'),
        ('adaptive', 'robust_fair'),
        ('adaptive', 'krum'),
    ]

    # Seeds to run
    seeds = [42, 100, 999]

    # Run experiments for each seed
    all_runs = []

    for seed in seeds:
        results = run_single_seed(seed, experiments)
        all_runs.append(results)

        # Print summary for this seed
        SummaryPrinter.print_single_seed_summary(results, seed)

        # Generate plots for this seed
        Visualizer.plot_single_seed_results(results, seed, Config.OUTPUT_DIR)

    # Generate comparison plots across all seeds
    print("\n" + "=" * 80)
    print("GENERATING CROSS-SEED COMPARISON PLOTS")
    print("=" * 80)

    Visualizer.plot_comparison_across_seeds(all_runs, seeds, Config.OUTPUT_DIR)
    Visualizer.plot_time_series_comparison(all_runs, seeds, Config.OUTPUT_DIR)

    # Print aggregate summary
    SummaryPrinter.print_multi_seed_summary(all_runs, seeds)

    print("\n" + "=" * 80)
    print("EXPERIMENT COMPLETE")
    print(f"Results saved to: {Config.OUTPUT_DIR}/")
    print("=" * 80)

    return all_runs


# ==============================================================================
# ADDITIONAL ANALYSIS FUNCTIONS
# ==============================================================================

def generate_heatmap(all_runs, seeds, save_dir):
    """Generate heatmap of attack/defense combinations"""

    # Collect mean metrics
    metrics = defaultdict(lambda: {'accuracy': [], 'dp': []})

    for run_results in all_runs:
        for (attack, defense), history in run_results.items():
            key = (attack, defense)
            metrics[key]['accuracy'].append(history['accuracy'][-1])
            metrics[key]['dp'].append(history['demographic_parity'][-1])

    # Get unique attacks and defenses
    attacks = sorted(set(k[0] for k in metrics.keys()))
    defenses = sorted(set(k[1] for k in metrics.keys()))

    # Create matrices
    acc_matrix = np.zeros((len(attacks), len(defenses)))
    dp_matrix = np.zeros((len(attacks), len(defenses)))

    for i, attack in enumerate(attacks):
        for j, defense in enumerate(defenses):
            key = (attack, defense)
            if key in metrics:
                acc_matrix[i, j] = np.mean(metrics[key]['accuracy'])
                dp_matrix[i, j] = np.mean(metrics[key]['dp'])
            else:
                acc_matrix[i, j] = np.nan
                dp_matrix[i, j] = np.nan

    # Create heatmaps
    fig, axes = plt.subplots(1, 2, figsize=(16, 8))
    fig.suptitle('Attack-Defense Performance Heatmaps', fontsize=16, fontweight='bold')

    # Accuracy heatmap
    ax = axes[0]
    im = ax.imshow(acc_matrix, cmap='RdYlGn', aspect='auto', vmin=0.5, vmax=1.0)
    ax.set_xticks(range(len(defenses)))
    ax.set_yticks(range(len(attacks)))
    ax.set_xticklabels(defenses, rotation=45, ha='right')
    ax.set_yticklabels(attacks)
    ax.set_title('Accuracy', fontsize=12)
    ax.set_xlabel('Defense')
    ax.set_ylabel('Attack')
    plt.colorbar(im, ax=ax, shrink=0.8)

    # Add text annotations
    for i in range(len(attacks)):
        for j in range(len(defenses)):
            if not np.isnan(acc_matrix[i, j]):
                ax.text(j, i, f'{acc_matrix[i, j]:.3f}',
                       ha='center', va='center', fontsize=8)

    # DP Gap heatmap
    ax = axes[1]
    im = ax.imshow(dp_matrix, cmap='RdYlGn_r', aspect='auto', vmin=0, vmax=0.5)
    ax.set_xticks(range(len(defenses)))
    ax.set_yticks(range(len(attacks)))
    ax.set_xticklabels(defenses, rotation=45, ha='right')
    ax.set_yticklabels(attacks)
    ax.set_title('Demographic Parity Gap (lower is better)', fontsize=12)
    ax.set_xlabel('Defense')
    ax.set_ylabel('Attack')
    plt.colorbar(im, ax=ax, shrink=0.8)

    # Add text annotations
    for i in range(len(attacks)):
        for j in range(len(defenses)):
            if not np.isnan(dp_matrix[i, j]):
                ax.text(j, i, f'{dp_matrix[i, j]:.3f}',
                       ha='center', va='center', fontsize=8)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    filename = os.path.join(save_dir, 'heatmap_attack_defense.png')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"[SAVED] {filename}")
    plt.close()


def generate_radar_chart(all_runs, seeds, save_dir):
    """Generate radar charts comparing defenses"""

    # Collect metrics for fairness_poison attack
    attack = 'fairness_poison'
    defenses = ['fedavg', 'median', 'krum', 'fairness_only', 'robust_fair']

    metrics = {defense: {'accuracy': [], 'dp': [], 'eo': []} for defense in defenses}

    for run_results in all_runs:
        for defense in defenses:
            key = (attack, defense)
            if key in run_results:
                history = run_results[key]
                metrics[defense]['accuracy'].append(history['accuracy'][-1])
                metrics[defense]['dp'].append(history['demographic_parity'][-1])
                metrics[defense]['eo'].append(history['equalized_odds'][-1])

    # Compute means
    categories = ['Accuracy', 'Fairness (1-DP)', 'Eq. Odds (1-EO)']
    N = len(categories)

    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))

    colors = plt.cm.Set1(np.linspace(0, 1, len(defenses)))

    for defense, color in zip(defenses, colors):
        if metrics[defense]['accuracy']:
            values = [
                np.mean(metrics[defense]['accuracy']),
                1 - np.mean(metrics[defense]['dp']),
                1 - np.mean(metrics[defense]['eo'])
            ]
            values += values[:1]

            ax.plot(angles, values, 'o-', linewidth=2, label=defense, color=color)
            ax.fill(angles, values, alpha=0.1, color=color)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=12)
    ax.set_ylim(0, 1)
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
    ax.set_title(f'Defense Performance Radar (vs {attack})', fontsize=14, fontweight='bold', pad=20)

    filename = os.path.join(save_dir, 'radar_defense_comparison.png')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"[SAVED] {filename}")
    plt.close()


# ==============================================================================
# RUN
# ==============================================================================

if __name__ == "__main__":
    all_runs = main()

    # Generate additional visualizations
    print("\nGenerating additional visualizations...")
    generate_heatmap(all_runs, [42, 100, 999], Config.OUTPUT_DIR)
    generate_radar_chart(all_runs, [42, 100, 999], Config.OUTPUT_DIR)

    print("\nAll visualizations complete!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
import random
import matplotlib.pyplot as plt
from collections import defaultdict
import warnings
import os

warnings.filterwarnings('ignore')

"""
================================================================================
EXPERIMENT 1: Synthetic Data - Multi-Seed Analysis
================================================================================
"""

# ==============================================================================
# CONFIGURATION
# ==============================================================================

class Config:
    """Configuration class for experiment parameters"""
    N_SAMPLES = 2000
    N_FEATURES = 10
    N_CLIENTS = 6
    N_ROUNDS = 20
    LOCAL_EPOCHS = 5
    LEARNING_RATE = 0.01
    ADVERSARY_INDICES = [4, 5]
    ATTACK_START_ROUND = 5
    SEED = 42

    # Output directory
    OUTPUT_DIR = "experiment_results"

    @classmethod
    def set_seed(cls, seed=None):
        """Set random seeds for reproducibility"""
        if seed is not None:
            cls.SEED = seed
        random.seed(cls.SEED)
        np.random.seed(cls.SEED)
        torch.manual_seed(cls.SEED)
        if torch.cuda.is_available():
            torch.cuda.manual_seed(cls.SEED)

    @classmethod
    def create_output_dir(cls):
        """Create output directory if it doesn't exist"""
        if not os.path.exists(cls.OUTPUT_DIR):
            os.makedirs(cls.OUTPUT_DIR)


# ==============================================================================
# DATA GENERATION
# ==============================================================================

class DataGenerator:
    """Class for generating and managing synthetic data"""

    @staticmethod
    def load_synthetic_data():
        """Generate synthetic classification data with sensitive attributes"""
        X, y = make_classification(
            n_samples=Config.N_SAMPLES,
            n_features=Config.N_FEATURES,
            n_informative=6,
            n_redundant=2,
            random_state=Config.SEED
        )

        # Generate sensitive attribute
        sensitive = np.random.randint(0, 2, size=Config.N_SAMPLES)

        # Introduce bias based on sensitive attribute
        for i in range(Config.N_SAMPLES):
            if sensitive[i] == 1 and random.random() < 0.3:
                y[i] = 1 - y[i]

        # Standardize features
        scaler = StandardScaler()
        X = scaler.fit_transform(X)

        return X, y, sensitive

    @staticmethod
    def create_clients(X, y, sensitive):
        """Split data into clients with heterogeneous bias rates"""
        n_samples = len(X)
        indices = np.arange(n_samples)
        np.random.shuffle(indices)
        splits = np.array_split(indices, Config.N_CLIENTS)

        clients = []
        for i, split in enumerate(splits):
            # Each client has different local bias rate
            local_bias = 0.1 + (i * 0.1)

            X_c = X[split].copy()
            y_c = y[split].copy()
            s_c = sensitive[split].copy()

            # Apply local bias
            for j in range(len(y_c)):
                if s_c[j] == 1 and random.random() < local_bias:
                    y_c[j] = 1 - y_c[j]

            clients.append({
                'id': i,
                'X': X_c,
                'y': y_c,
                'sensitive': s_c,
                'bias_rate': local_bias
            })

        return clients


# ==============================================================================
# MODEL DEFINITION
# ==============================================================================

class FederatedModel(nn.Module):
    """Neural network model for federated learning"""

    def __init__(self, input_dim=10):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.net(x)

    def predict_proba(self, x):
        """Get probability predictions"""
        with torch.no_grad():
            logits = self.forward(x)
            return torch.sigmoid(logits)


# ==============================================================================
# FAIRNESS METRICS
# ==============================================================================

class FairnessMetrics:
    """Class for computing fairness metrics"""

    @staticmethod
    def demographic_parity(preds, sensitive):
        """Compute demographic parity gap"""
        preds = np.asarray(preds).flatten()
        binary_preds = (preds > 0.5).astype(int)

        rate_0 = binary_preds[sensitive == 0].mean() if np.any(sensitive == 0) else 0.5
        rate_1 = binary_preds[sensitive == 1].mean() if np.any(sensitive == 1) else 0.5

        return abs(rate_0 - rate_1)

    @staticmethod
    def equalized_odds(preds, labels, sensitive):
        """Compute equalized odds gap"""
        preds = np.asarray(preds).flatten()
        binary_preds = (preds > 0.5).astype(int)

        mask_0_pos = (sensitive == 0) & (labels == 1)
        mask_1_pos = (sensitive == 1) & (labels == 1)

        tpr_0 = binary_preds[mask_0_pos].mean() if np.any(mask_0_pos) else 0.5
        tpr_1 = binary_preds[mask_1_pos].mean() if np.any(mask_1_pos) else 0.5

        return abs(tpr_0 - tpr_1)


# ==============================================================================
# TRAINING UTILITIES
# ==============================================================================

class TrainingUtils:
    """Utility functions for model training"""

    @staticmethod
    def safe_train(model, X, y, epochs=5, lr=0.01):
        """Safe training with BCEWithLogitsLoss"""
        optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X_t = torch.tensor(X, dtype=torch.float32)
        y_t = torch.tensor(y, dtype=torch.float32).view(-1, 1)

        model.train()
        for _ in range(epochs):
            optimizer.zero_grad()
            logits = model(X_t)
            loss = loss_fn(logits, y_t)

            if torch.isnan(loss) or torch.isinf(loss):
                break

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

        return model.state_dict()

    @staticmethod
    def honest_train(global_weights, client, input_dim):
        """Honest client training"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)
        return TrainingUtils.safe_train(
            model, client['X'], client['y'],
            Config.LOCAL_EPOCHS, Config.LEARNING_RATE
        )


# ==============================================================================
# ATTACK IMPLEMENTATIONS
# ==============================================================================

class Attacks:
    """Collection of attack strategies"""

    @staticmethod
    def none(global_weights, client, input_dim):
        """No attack - honest training"""
        return TrainingUtils.honest_train(global_weights, client, input_dim)

    @staticmethod
    def random_weights(global_weights, client, input_dim):
        """Random weight attack"""
        return FederatedModel(input_dim=input_dim).state_dict()

    @staticmethod
    def fairness_poison(global_weights, client, input_dim):
        """Fairness poisoning attack"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        # Create adversarial labels that discriminate based on sensitive attribute
        adversarial_y = np.zeros(len(client['y']))
        adversarial_y[client['sensitive'] == 0] = client['y'][client['sensitive'] == 0]

        return TrainingUtils.safe_train(model, client['X'], adversarial_y, epochs=15, lr=0.05)

    @staticmethod
    def label_flip(global_weights, client, input_dim):
        """Label flipping attack"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        flipped_y = client['y'].copy()
        for i in range(len(flipped_y)):
            if client['sensitive'][i] == 1 and random.random() < 0.6:
                flipped_y[i] = 1 - flipped_y[i]

        return TrainingUtils.safe_train(
            model, client['X'], flipped_y,
            Config.LOCAL_EPOCHS, Config.LEARNING_RATE
        )

    @staticmethod
    def gradient_scale(global_weights, client, input_dim):
        """Gradient scaling attack"""
        honest = TrainingUtils.honest_train(global_weights, client, input_dim)

        scaled = {}
        for key in honest.keys():
            diff = honest[key] - global_weights[key]
            scaled[key] = global_weights[key] + 2.0 * torch.clamp(diff, -0.5, 0.5)

        return scaled

    @staticmethod
    def fairness_spoof(global_weights, client, input_dim):
        """Fairness spoofing attack - appears fair locally but introduces global bias"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        optimizer = optim.Adam(model.parameters(), lr=0.005, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X = torch.tensor(client['X'], dtype=torch.float32)
        y = torch.tensor(client['y'], dtype=torch.float32).view(-1, 1)
        sensitive = client['sensitive']

        mask_0 = torch.tensor(sensitive == 0)
        mask_1 = torch.tensor(sensitive == 1)

        model.train()
        for _ in range(15):
            optimizer.zero_grad()

            logits = model(X)
            probs = torch.sigmoid(logits)

            # Classification loss
            ce_loss = loss_fn(logits, y)

            # Local fairness matching (appears fair locally)
            if mask_0.any() and mask_1.any():
                fair_loss = (probs[mask_0].mean() - probs[mask_1].mean()) ** 2
            else:
                fair_loss = torch.tensor(0.0)

            # Hidden bias towards group 1
            if mask_1.any():
                bias_loss = 0.05 * probs[mask_1].mean()
            else:
                bias_loss = torch.tensor(0.0)

            total_loss = ce_loss + 0.2 * fair_loss + bias_loss

            if torch.isnan(total_loss):
                break

            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)
            optimizer.step()

        return model.state_dict()

    @staticmethod
    def adaptive(global_weights, client, input_dim):
        """Adaptive attack designed to evade detection"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        optimizer = optim.Adam(model.parameters(), lr=0.003, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X = torch.tensor(client['X'], dtype=torch.float32)
        y = torch.tensor(client['y'], dtype=torch.float32).view(-1, 1)
        sensitive = client['sensitive']

        mask_0 = torch.tensor(sensitive == 0)
        mask_1 = torch.tensor(sensitive == 1)

        initial_weights = {k: v.clone() for k, v in model.state_dict().items()}

        model.train()
        for _ in range(20):
            optimizer.zero_grad()

            logits = model(X)
            probs = torch.sigmoid(logits)

            ce_loss = loss_fn(logits, y)

            # Stay fair locally
            if mask_0.any() and mask_1.any():
                fair_loss = torch.abs(probs[mask_0].mean() - probs[mask_1].mean())
            else:
                fair_loss = torch.tensor(0.0)

            # Stay close to global weights (evade distance-based detection)
            weight_dist = sum(
                torch.norm(model.state_dict()[k].float() - initial_weights[k].float())
                for k in initial_weights.keys()
            )

            # Subtle bias
            if mask_1.any():
                bias_loss = -0.02 * ((probs[mask_1] - 0.5) ** 2).mean()
            else:
                bias_loss = torch.tensor(0.0)

            total_loss = ce_loss + 0.3 * fair_loss + 0.002 * weight_dist + bias_loss

            if torch.isnan(total_loss):
                break

            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.3)
            optimizer.step()

        return model.state_dict()


# ==============================================================================
# AGGREGATION METHODS (DEFENSES)
# ==============================================================================

class Aggregators:
    """Collection of aggregation/defense strategies"""

    @staticmethod
    def fedavg(weights_list, **kwargs):
        """Federated Averaging"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            agg[key] = torch.mean(stacked, dim=0)
        return agg, []

    @staticmethod
    def median(weights_list, **kwargs):
        """Coordinate-wise median aggregation"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            agg[key] = torch.median(stacked, dim=0)[0]
        return agg, []

    @staticmethod
    def trimmed_mean(weights_list, trim=1, **kwargs):
        """Trimmed mean aggregation"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            sorted_w, _ = torch.sort(stacked, dim=0)
            if trim > 0 and len(weights_list) > 2 * trim:
                trimmed = sorted_w[trim:-trim]
            else:
                trimmed = sorted_w
            agg[key] = torch.mean(trimmed, dim=0)
        return agg, []

    @staticmethod
    def krum(weights_list, n_bad=2, **kwargs):
        """Krum aggregation"""
        n = len(weights_list)
        n_select = max(1, n - n_bad)

        # Compute pairwise distances
        dists = np.zeros((n, n))
        for i in range(n):
            for j in range(i + 1, n):
                d = sum(
                    torch.norm(weights_list[i][k].float() - weights_list[j][k].float()).item() ** 2
                    for k in weights_list[i].keys()
                )
                dists[i, j] = dists[j, i] = np.sqrt(d)

        # Compute scores
        scores = [np.sum(np.sort(dists[i])[1:n_select]) for i in range(n)]
        selected = np.argsort(scores)[:n_select]

        # Aggregate selected
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([weights_list[i][key].float() for i in selected])
            agg[key] = torch.mean(stacked, dim=0)

        excluded = [i for i in range(n) if i not in selected]
        return agg, excluded

    @staticmethod
    def fairness_only(weights_list, clients, input_dim, **kwargs):
        """Fairness-weighted aggregation"""
        fairness_scores = []

        for w, c in zip(weights_list, clients):
            model = FederatedModel(input_dim=input_dim)
            model.load_state_dict(w)
            model.eval()

            with torch.no_grad():
                preds = model.predict_proba(
                    torch.tensor(c['X'], dtype=torch.float32)
                ).numpy()

            dp = FairnessMetrics.demographic_parity(preds, c['sensitive'])
            fairness_scores.append(dp)

        # Lower DP gap = higher weight
        weights = 1.0 / (np.array(fairness_scores) + 0.01)
        weights = weights / weights.sum()

        agg = {}
        for key in weights_list[0].keys():
            weighted_sum = torch.zeros_like(weights_list[0][key].float())
            for i, w in enumerate(weights_list):
                weighted_sum += weights[i] * w[key].float()
            agg[key] = weighted_sum

        return agg, []

    @staticmethod
    def robust_fair(weights_list, clients, global_weights, X_all, s_all, input_dim, **kwargs):
        """Robust fairness-aware aggregation"""
        n = len(weights_list)

        # Compute distance scores
        distances = []
        for w in weights_list:
            d = sum(
                torch.norm(w[k].float() - global_weights[k].float()).item() ** 2
                for k in global_weights.keys()
            )
            distances.append(np.sqrt(d))

        distances = np.array(distances)
        z_scores = (distances - distances.mean()) / (distances.std() + 1e-6)

        # Compute fairness consistency
        consistency = []
        for w, c in zip(weights_list, clients):
            model = FederatedModel(input_dim=input_dim)
            model.load_state_dict(w)
            model.eval()

            with torch.no_grad():
                local_preds = model.predict_proba(
                    torch.tensor(c['X'], dtype=torch.float32)
                ).numpy()
                global_preds = model.predict_proba(
                    torch.tensor(X_all, dtype=torch.float32)
                ).numpy()

            local_dp = FairnessMetrics.demographic_parity(local_preds, c['sensitive'])
            global_dp = FairnessMetrics.demographic_parity(global_preds, s_all)
            consistency.append(abs(local_dp - global_dp))

        consistency = np.array(consistency)

        # Compute trust scores
        trust = []
        suspicious = []
        for i in range(n):
            t = 1.0 / (1.0 + abs(z_scores[i]) + consistency[i])
            if z_scores[i] > 2.0 or consistency[i] > 0.3:
                suspicious.append(i)
                t *= 0.1
            trust.append(t)

        trust = np.array(trust)
        trust = trust / trust.sum()

        # Weighted aggregation
        agg = {}
        for key in weights_list[0].keys():
            weighted_sum = torch.zeros_like(weights_list[0][key].float())
            for i, w in enumerate(weights_list):
                weighted_sum += trust[i] * w[key].float()
            agg[key] = weighted_sum

        return agg, suspicious


# ==============================================================================
# EXPERIMENT RUNNER
# ==============================================================================

class ExperimentRunner:
    """Class for running federated learning experiments"""

    # Attack function mapping
    ATTACKS = {
        'none': Attacks.none,
        'random': Attacks.random_weights,
        'fairness_poison': Attacks.fairness_poison,
        'label_flip': Attacks.label_flip,
        'gradient_scale': Attacks.gradient_scale,
        'fairness_spoof': Attacks.fairness_spoof,
        'adaptive': Attacks.adaptive
    }

    # Defense function mapping
    DEFENSES = {
        'fedavg': Aggregators.fedavg,
        'median': Aggregators.median,
        'trimmed': Aggregators.trimmed_mean,
        'krum': Aggregators.krum,
        'fairness_only': Aggregators.fairness_only,
        'robust_fair': Aggregators.robust_fair
    }

    @staticmethod
    def run_single_experiment(attack_name, defense_name, X, y, sensitive, clients, input_dim):
        """Run a single experiment configuration"""
        attack_fn = ExperimentRunner.ATTACKS[attack_name]
        agg_fn = ExperimentRunner.DEFENSES[defense_name]

        global_model = FederatedModel(input_dim=input_dim)
        history = defaultdict(list)

        for round_num in range(Config.N_ROUNDS):
            attack_active = round_num >= Config.ATTACK_START_ROUND
            client_weights = []

            # Local training
            for i, client in enumerate(clients):
                is_adversary = i in Config.ADVERSARY_INDICES and attack_active

                try:
                    if is_adversary:
                        w = attack_fn(global_model.state_dict(), client, input_dim)
                    else:
                        w = TrainingUtils.honest_train(
                            global_model.state_dict(), client, input_dim
                        )
                except Exception:
                    w = TrainingUtils.honest_train(
                        global_model.state_dict(), client, input_dim
                    )

                client_weights.append(w)

            # Aggregation
            try:
                agg_weights, detected = agg_fn(
                    client_weights,
                    clients=clients,
                    global_weights=global_model.state_dict(),
                    X_all=X,
                    s_all=sensitive,
                    input_dim=input_dim
                )
                global_model.load_state_dict(agg_weights)
            except Exception:
                agg_weights, _ = Aggregators.fedavg(client_weights)
                global_model.load_state_dict(agg_weights)

            # Evaluation
            global_model.eval()
            with torch.no_grad():
                preds = global_model.predict_proba(
                    torch.tensor(X, dtype=torch.float32)
                ).numpy()

            acc = accuracy_score(y, (preds > 0.5).astype(int))
            dp = FairnessMetrics.demographic_parity(preds, sensitive)
            eo = FairnessMetrics.equalized_odds(preds, y, sensitive)

            history['accuracy'].append(acc)
            history['demographic_parity'].append(dp)
            history['equalized_odds'].append(eo)
            history['round'].append(round_num + 1)

        return dict(history)


# ==============================================================================
# VISUALIZATION
# ==============================================================================

class Visualizer:
    """Class for generating experiment visualizations"""

    # Color schemes
    ATTACK_COLORS = {
        'none': 'green',
        'random': 'orange',
        'fairness_poison': 'red',
        'label_flip': 'purple',
        'gradient_scale': 'blue',
        'fairness_spoof': 'magenta',
        'adaptive': 'black'
    }

    DEFENSE_MARKERS = {
        'fedavg': 'o',
        'median': 's',
        'trimmed': 'd',
        'krum': 'p',
        'fairness_only': '^',
        'robust_fair': '*'
    }

    @staticmethod
    def plot_single_seed_results(results, seed, save_dir):
        """Generate comprehensive plots for a single seed"""
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle(f'Federated Learning Experiment Results - Seed {seed}', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy over rounds (attacks vs fedavg)
        ax = axes[0, 0]
        attacks = ['none', 'random', 'fairness_poison', 'fairness_spoof', 'adaptive']

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['accuracy'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7, label='Attack Start')
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy Under Different Attacks (vs FedAvg)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)
        ax.set_ylim([0.4, 1.0])

        # Plot 2: Demographic Parity over rounds (attacks vs fedavg)
        ax = axes[0, 1]

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Fairness Under Different Attacks (vs FedAvg)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 3: Equalized Odds over rounds
        ax = axes[0, 2]

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['equalized_odds'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Equalized Odds Gap', fontsize=11)
        ax.set_title('Equalized Odds Under Different Attacks', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 4: Defense comparison (vs fairness_poison)
        ax = axes[1, 0]
        defenses = ['fedavg', 'median', 'trimmed', 'krum', 'fairness_only', 'robust_fair']

        for defense in defenses:
            key = ('fairness_poison', defense)
            if key in results:
                ls = '--' if defense == 'fedavg' else '-'
                lw = 3 if defense in ['robust_fair', 'fedavg'] else 2
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=defense, linestyle=ls, linewidth=lw, marker='s', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Defense Effectiveness vs Fairness Poisoning', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 5: Defense comparison (vs adaptive)
        ax = axes[1, 1]

        for defense in defenses:
            key = ('adaptive', defense)
            if key in results:
                ls = '--' if defense == 'fedavg' else '-'
                lw = 3 if defense in ['robust_fair', 'fedavg'] else 2
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=defense, linestyle=ls, linewidth=lw, marker='s', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Defense Effectiveness vs Adaptive Attack', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 6: Accuracy vs Fairness tradeoff (scatter)
        ax = axes[1, 2]

        for (attack, defense), history in results.items():
            color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
            marker = Visualizer.DEFENSE_MARKERS.get(defense, 'o')

            ax.scatter(
                history['accuracy'][-1],
                history['demographic_parity'][-1],
                c=color, marker=marker, s=150, alpha=0.7,
                edgecolors='black', linewidth=0.5,
                label=f"{attack[:6]}/{defense[:6]}"
            )

        ax.set_xlabel('Final Accuracy', fontsize=11)
        ax.set_ylabel('Final DP Gap', fontsize=11)
        ax.set_title('Accuracy-Fairness Tradeoff', fontsize=12)
        ax.grid(True, alpha=0.3)

        # Add legend for color/marker explanation
        ax.text(0.02, 0.98, 'Colors = Attacks\nMarkers = Defenses',
               transform=ax.transAxes, fontsize=9, verticalalignment='top',
               bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, f'results_seed_{seed}.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()

    @staticmethod
    def plot_comparison_across_seeds(all_runs, seeds, save_dir):
        """Generate comparison plots across all seeds"""

        # Collect metrics for each experiment across seeds
        metrics_by_exp = defaultdict(lambda: {'accuracy': [], 'dp': [], 'eo': []})

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                metrics_by_exp[key]['accuracy'].append(history['accuracy'][-1])
                metrics_by_exp[key]['dp'].append(history['demographic_parity'][-1])
                metrics_by_exp[key]['eo'].append(history['equalized_odds'][-1])

        # Create comparison figure
        fig, axes = plt.subplots(2, 2, figsize=(16, 14))
        fig.suptitle(f'Multi-Seed Comparison (Seeds: {seeds})', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy comparison bar chart
        ax = axes[0, 0]
        experiments = list(metrics_by_exp.keys())
        x = np.arange(len(experiments))

        means = [np.mean(metrics_by_exp[exp]['accuracy']) for exp in experiments]
        stds = [np.std(metrics_by_exp[exp]['accuracy']) for exp in experiments]

        colors = [Visualizer.ATTACK_COLORS.get(exp[0], 'gray') for exp in experiments]
        bars = ax.bar(x, means, yerr=stds, capsize=3, color=colors, alpha=0.7, edgecolor='black')

        ax.set_xlabel('Experiment', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Final Accuracy (Mean ± Std across Seeds)', fontsize=12)
        ax.set_xticks(x)
        ax.set_xticklabels([f"{a[:4]}/{d[:4]}" for a, d in experiments], rotation=45, ha='right', fontsize=8)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 2: DP comparison bar chart
        ax = axes[0, 1]

        means = [np.mean(metrics_by_exp[exp]['dp']) for exp in experiments]
        stds = [np.std(metrics_by_exp[exp]['dp']) for exp in experiments]

        bars = ax.bar(x, means, yerr=stds, capsize=3, color=colors, alpha=0.7, edgecolor='black')

        ax.set_xlabel('Experiment', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Final DP Gap (Mean ± Std across Seeds)', fontsize=12)
        ax.set_xticks(x)
        ax.set_xticklabels([f"{a[:4]}/{d[:4]}" for a, d in experiments], rotation=45, ha='right', fontsize=8)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 3: Box plot for accuracy by attack type
        ax = axes[1, 0]

        attack_types = ['none', 'fairness_poison', 'fairness_spoof', 'adaptive']
        defense = 'fedavg'

        box_data = []
        labels = []
        for attack in attack_types:
            key = (attack, defense)
            if key in metrics_by_exp:
                box_data.append(metrics_by_exp[key]['accuracy'])
                labels.append(attack)

        bp = ax.boxplot(box_data, labels=labels, patch_artist=True)
        for patch, attack in zip(bp['boxes'], attack_types):
            patch.set_facecolor(Visualizer.ATTACK_COLORS.get(attack, 'gray'))
            patch.set_alpha(0.7)

        ax.set_xlabel('Attack Type', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy Distribution by Attack (vs FedAvg)', fontsize=12)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 4: Box plot for DP gap by defense type
        ax = axes[1, 1]

        attack = 'fairness_poison'
        defense_types = ['fedavg', 'median', 'krum', 'fairness_only', 'robust_fair']

        box_data = []
        labels = []
        for defense in defense_types:
            key = (attack, defense)
            if key in metrics_by_exp:
                box_data.append(metrics_by_exp[key]['dp'])
                labels.append(defense)

        bp = ax.boxplot(box_data, labels=labels, patch_artist=True)
        colors = plt.cm.Set2(np.linspace(0, 1, len(labels)))
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)

        ax.set_xlabel('Defense Type', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap Distribution by Defense (vs Fairness Poison)', fontsize=12)
        ax.grid(True, alpha=0.3, axis='y')

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, 'comparison_across_seeds.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()

    @staticmethod
    def plot_time_series_comparison(all_runs, seeds, save_dir):
        """Plot time series with confidence bands across seeds"""

        # Collect time series data
        time_series = defaultdict(lambda: defaultdict(list))

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                time_series[key]['accuracy'].append(history['accuracy'])
                time_series[key]['dp'].append(history['demographic_parity'])

        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle(f'Time Series with Confidence Bands (Seeds: {seeds})', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy for different attacks
        ax = axes[0, 0]
        attacks = ['none', 'fairness_poison', 'fairness_spoof', 'adaptive']
        defense = 'fedavg'

        for attack in attacks:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['accuracy'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(rounds, mean, label=attack, color=color, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, color=color, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy with Confidence Bands (Attacks vs FedAvg)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 2: DP for different attacks
        ax = axes[0, 1]

        for attack in attacks:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['dp'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(rounds, mean, label=attack, color=color, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, color=color, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap with Confidence Bands (Attacks vs FedAvg)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 3: Accuracy for different defenses
        ax = axes[1, 0]
        attack = 'fairness_poison'
        defenses = ['fedavg', 'median', 'krum', 'robust_fair']

        for defense in defenses:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['accuracy'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                ax.plot(rounds, mean, label=defense, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy with Confidence Bands (Defenses vs Fairness Poison)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 4: DP for different defenses
        ax = axes[1, 1]

        for defense in defenses:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['dp'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                ax.plot(rounds, mean, label=defense, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap with Confidence Bands (Defenses vs Fairness Poison)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, 'time_series_comparison.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()


# ==============================================================================
# SUMMARY PRINTER
# ==============================================================================

class SummaryPrinter:
    """Class for printing experiment summaries"""

    @staticmethod
    def print_single_seed_summary(results, seed):
        """Print summary for a single seed"""
        print("\n" + "=" * 80)
        print(f"SUMMARY TABLE - SEED {seed}")
        print("=" * 80)
        print(f"{'Attack':<20} {'Defense':<15} {'Accuracy':>10} {'DP Gap':>10} {'EO Gap':>10}")
        print("-" * 80)

        for (attack, defense) in sorted(results.keys()):
            h = results[(attack, defense)]
            print(f"{attack:<20} {defense:<15} {h['accuracy'][-1]:>10.4f} "
                  f"{h['demographic_parity'][-1]:>10.4f} {h['equalized_odds'][-1]:>10.4f}")

        print("=" * 80)

    @staticmethod
    def print_multi_seed_summary(all_runs, seeds):
        """Print aggregate summary across all seeds"""
        print("\n" + "=" * 80)
        print(f"AGGREGATE SUMMARY ACROSS SEEDS: {seeds}")
        print("=" * 80)

        # Collect all metrics
        metrics = defaultdict(lambda: {'accuracy': [], 'dp': [], 'eo': []})

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                metrics[key]['accuracy'].append(history['accuracy'][-1])
                metrics[key]['dp'].append(history['demographic_parity'][-1])
                metrics[key]['eo'].append(history['equalized_odds'][-1])

        print(f"{'Attack':<18} {'Defense':<12} {'Acc (mean±std)':>16} {'DP (mean±std)':>16} {'EO (mean±std)':>16}")
        print("-" * 80)

        for (attack, defense) in sorted(metrics.keys()):
            m = metrics[(attack, defense)]
            acc_mean, acc_std = np.mean(m['accuracy']), np.std(m['accuracy'])
            dp_mean, dp_std = np.mean(m['dp']), np.std(m['dp'])
            eo_mean, eo_std = np.mean(m['eo']), np.std(m['eo'])

            print(f"{attack:<18} {defense:<12} {acc_mean:>7.4f}±{acc_std:<6.4f} "
                  f"{dp_mean:>7.4f}±{dp_std:<6.4f} {eo_mean:>7.4f}±{eo_std:<6.4f}")

        print("=" * 80)

        # Key findings
        print("\n" + "=" * 80)
        print("KEY FINDINGS")
        print("=" * 80)

        # Baseline comparison
        baseline_key = ('none', 'fedavg')
        if baseline_key in metrics:
            baseline_dp = np.mean(metrics[baseline_key]['dp'])
            baseline_acc = np.mean(metrics[baseline_key]['accuracy'])
            print(f"\nBaseline (no attack, fedavg):")
            print(f"  Accuracy: {baseline_acc:.4f}")
            print(f"  DP Gap: {baseline_dp:.4f}")

            print("\nImpact of attacks (vs baseline):")
            for attack in ['fairness_poison', 'fairness_spoof', 'adaptive']:
                key = (attack, 'fedavg')
                if key in metrics:
                    attack_dp = np.mean(metrics[key]['dp'])
                    attack_acc = np.mean(metrics[key]['accuracy'])
                    print(f"  {attack}:")
                    print(f"    Accuracy: {attack_acc:.4f} (Δ = {attack_acc - baseline_acc:+.4f})")
                    print(f"    DP Gap: {attack_dp:.4f} (Δ = {attack_dp - baseline_dp:+.4f})")

        # Defense comparison
        print("\nDefense effectiveness (vs fairness_poison attack):")
        attack = 'fairness_poison'
        baseline_defense = 'fedavg'

        if (attack, baseline_defense) in metrics:
            baseline_dp = np.mean(metrics[(attack, baseline_defense)]['dp'])

            for defense in ['median', 'trimmed', 'krum', 'fairness_only', 'robust_fair']:
                key = (attack, defense)
                if key in metrics:
                    defense_dp = np.mean(metrics[key]['dp'])
                    improvement = ((baseline_dp - defense_dp) / baseline_dp) * 100 if baseline_dp > 0 else 0
                    print(f"  {defense}: DP = {defense_dp:.4f} ({improvement:+.1f}% improvement)")

        print("=" * 80)


# ==============================================================================
# MAIN EXECUTION
# ==============================================================================

def run_single_seed(seed, experiments):
    """Run all experiments for a single seed"""
    print(f"\n{'='*80}")
    print(f"RUNNING EXPERIMENTS FOR SEED {seed}")
    print(f"{'='*80}")

    # Set seed and load data
    Config.set_seed(seed)
    X, y, sensitive = DataGenerator.load_synthetic_data()
    clients = DataGenerator.create_clients(X, y, sensitive)
    input_dim = X.shape[1]

    print(f"Data: {len(X)} samples, {input_dim} features")
    print(f"Clients: {len(clients)}, Adversaries: {Config.ADVERSARY_INDICES}")
    print()

    results = {}

    for idx, (attack, defense) in enumerate(experiments):
        print(f"[{idx+1:2d}/{len(experiments)}] {attack:<20} vs {defense:<15}", end=" ")

        try:
            history = ExperimentRunner.run_single_experiment(
                attack, defense, X, y, sensitive, clients, input_dim
            )
            results[(attack, defense)] = history
            print(f"| Acc: {history['accuracy'][-1]:.3f} | DP: {history['demographic_parity'][-1]:.3f}")
        except Exception as e:
            print(f"| ERROR: {str(e)[:40]}")

    return results


def main():
    """Main execution function"""
    print("=" * 80)
    print("FEDERATED LEARNING FAIRNESS EXPERIMENT")
    print("Multi-Seed Analysis with Separate Visualizations")
    print("=" * 80)

    # Create output directory
    Config.create_output_dir()

    # Define experiments
    experiments = [
        # Baselines
        ('none', 'fedavg'),
        ('none', 'median'),
        ('none', 'fairness_only'),
        ('none', 'robust_fair'),

        # Attacks vs FedAvg
        ('random', 'fedavg'),
        ('fairness_poison', 'fedavg'),
        ('label_flip', 'fedavg'),
        ('gradient_scale', 'fedavg'),
        ('fairness_spoof', 'fedavg'),
        ('adaptive', 'fedavg'),

        # Defenses vs fairness_poison
        ('fairness_poison', 'median'),
        ('fairness_poison', 'trimmed'),
        ('fairness_poison', 'krum'),
        ('fairness_poison', 'fairness_only'),
        ('fairness_poison', 'robust_fair'),

        # Defenses vs fairness_spoof
        ('fairness_spoof', 'median'),
        ('fairness_spoof', 'robust_fair'),

        # Defenses vs adaptive
        ('adaptive', 'median'),
        ('adaptive', 'fairness_only'),
        ('adaptive', 'robust_fair'),
        ('adaptive', 'krum'),
    ]

    # Seeds to run
    seeds = [42, 100, 999]

    # Run experiments for each seed
    all_runs = []

    for seed in seeds:
        results = run_single_seed(seed, experiments)
        all_runs.append(results)

        # Print summary for this seed
        SummaryPrinter.print_single_seed_summary(results, seed)

        # Generate plots for this seed
        Visualizer.plot_single_seed_results(results, seed, Config.OUTPUT_DIR)

    # Generate comparison plots across all seeds
    print("\n" + "=" * 80)
    print("GENERATING CROSS-SEED COMPARISON PLOTS")
    print("=" * 80)

    Visualizer.plot_comparison_across_seeds(all_runs, seeds, Config.OUTPUT_DIR)
    Visualizer.plot_time_series_comparison(all_runs, seeds, Config.OUTPUT_DIR)

    # Print aggregate summary
    SummaryPrinter.print_multi_seed_summary(all_runs, seeds)

    print("\n" + "=" * 80)
    print("EXPERIMENT COMPLETE")
    print(f"Results saved to: {Config.OUTPUT_DIR}/")
    print("=" * 80)

    return all_runs


# ==============================================================================
# ADDITIONAL ANALYSIS FUNCTIONS
# ==============================================================================

def generate_heatmap(all_runs, seeds, save_dir):
    """Generate heatmap of attack/defense combinations"""

    # Collect mean metrics
    metrics = defaultdict(lambda: {'accuracy': [], 'dp': []})

    for run_results in all_runs:
        for (attack, defense), history in run_results.items():
            key = (attack, defense)
            metrics[key]['accuracy'].append(history['accuracy'][-1])
            metrics[key]['dp'].append(history['demographic_parity'][-1])

    # Get unique attacks and defenses
    attacks = sorted(set(k[0] for k in metrics.keys()))
    defenses = sorted(set(k[1] for k in metrics.keys()))

    # Create matrices
    acc_matrix = np.zeros((len(attacks), len(defenses)))
    dp_matrix = np.zeros((len(attacks), len(defenses)))

    for i, attack in enumerate(attacks):
        for j, defense in enumerate(defenses):
            key = (attack, defense)
            if key in metrics:
                acc_matrix[i, j] = np.mean(metrics[key]['accuracy'])
                dp_matrix[i, j] = np.mean(metrics[key]['dp'])
            else:
                acc_matrix[i, j] = np.nan
                dp_matrix[i, j] = np.nan

    # Create heatmaps
    fig, axes = plt.subplots(1, 2, figsize=(16, 8))
    fig.suptitle('Attack-Defense Performance Heatmaps', fontsize=16, fontweight='bold')

    # Accuracy heatmap
    ax = axes[0]
    im = ax.imshow(acc_matrix, cmap='RdYlGn', aspect='auto', vmin=0.5, vmax=1.0)
    ax.set_xticks(range(len(defenses)))
    ax.set_yticks(range(len(attacks)))
    ax.set_xticklabels(defenses, rotation=45, ha='right')
    ax.set_yticklabels(attacks)
    ax.set_title('Accuracy', fontsize=12)
    ax.set_xlabel('Defense')
    ax.set_ylabel('Attack')
    plt.colorbar(im, ax=ax, shrink=0.8)

    # Add text annotations
    for i in range(len(attacks)):
        for j in range(len(defenses)):
            if not np.isnan(acc_matrix[i, j]):
                ax.text(j, i, f'{acc_matrix[i, j]:.3f}',
                       ha='center', va='center', fontsize=8)

    # DP Gap heatmap
    ax = axes[1]
    im = ax.imshow(dp_matrix, cmap='RdYlGn_r', aspect='auto', vmin=0, vmax=0.5)
    ax.set_xticks(range(len(defenses)))
    ax.set_yticks(range(len(attacks)))
    ax.set_xticklabels(defenses, rotation=45, ha='right')
    ax.set_yticklabels(attacks)
    ax.set_title('Demographic Parity Gap (lower is better)', fontsize=12)
    ax.set_xlabel('Defense')
    ax.set_ylabel('Attack')
    plt.colorbar(im, ax=ax, shrink=0.8)

    # Add text annotations
    for i in range(len(attacks)):
        for j in range(len(defenses)):
            if not np.isnan(dp_matrix[i, j]):
                ax.text(j, i, f'{dp_matrix[i, j]:.3f}',
                       ha='center', va='center', fontsize=8)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    filename = os.path.join(save_dir, 'heatmap_attack_defense.png')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"[SAVED] {filename}")
    plt.close()


def generate_radar_chart(all_runs, seeds, save_dir):
    """Generate radar charts comparing defenses"""

    # Collect metrics for fairness_poison attack
    attack = 'fairness_poison'
    defenses = ['fedavg', 'median', 'krum', 'fairness_only', 'robust_fair']

    metrics = {defense: {'accuracy': [], 'dp': [], 'eo': []} for defense in defenses}

    for run_results in all_runs:
        for defense in defenses:
            key = (attack, defense)
            if key in run_results:
                history = run_results[key]
                metrics[defense]['accuracy'].append(history['accuracy'][-1])
                metrics[defense]['dp'].append(history['demographic_parity'][-1])
                metrics[defense]['eo'].append(history['equalized_odds'][-1])

    # Compute means
    categories = ['Accuracy', 'Fairness (1-DP)', 'Eq. Odds (1-EO)']
    N = len(categories)

    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))

    colors = plt.cm.Set1(np.linspace(0, 1, len(defenses)))

    for defense, color in zip(defenses, colors):
        if metrics[defense]['accuracy']:
            values = [
                np.mean(metrics[defense]['accuracy']),
                1 - np.mean(metrics[defense]['dp']),
                1 - np.mean(metrics[defense]['eo'])
            ]
            values += values[:1]

            ax.plot(angles, values, 'o-', linewidth=2, label=defense, color=color)
            ax.fill(angles, values, alpha=0.1, color=color)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=12)
    ax.set_ylim(0, 1)
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
    ax.set_title(f'Defense Performance Radar (vs {attack})', fontsize=14, fontweight='bold', pad=20)

    filename = os.path.join(save_dir, 'radar_defense_comparison.png')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"[SAVED] {filename}")
    plt.close()


# ==============================================================================
# RUN
# ==============================================================================

if __name__ == "__main__":
    all_runs = main()

    # Generate additional visualizations
    print("\nGenerating additional visualizations...")
    generate_heatmap(all_runs, [42, 100, 999], Config.OUTPUT_DIR)
    generate_radar_chart(all_runs, [42, 100, 999], Config.OUTPUT_DIR)

    print("\nAll visualizations complete!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
import random
import matplotlib.pyplot as plt
from collections import defaultdict
import warnings
import os

warnings.filterwarnings('ignore')

"""
================================================================================
EXPERIMENT 1: Synthetic Data - Multi-Seed Analysis
================================================================================
"""

# ==============================================================================
# CONFIGURATION
# ==============================================================================

class Config:
    """Configuration class for experiment parameters"""
    N_SAMPLES = 2000
    N_FEATURES = 10
    N_CLIENTS = 6
    N_ROUNDS = 20
    LOCAL_EPOCHS = 5
    LEARNING_RATE = 0.01
    ADVERSARY_INDICES = [4, 5]
    ATTACK_START_ROUND = 5
    SEED = 42

    # Output directory
    OUTPUT_DIR = "experiment_results"

    @classmethod
    def set_seed(cls, seed=None):
        """Set random seeds for reproducibility"""
        if seed is not None:
            cls.SEED = seed
        random.seed(cls.SEED)
        np.random.seed(cls.SEED)
        torch.manual_seed(cls.SEED)
        if torch.cuda.is_available():
            torch.cuda.manual_seed(cls.SEED)

    @classmethod
    def create_output_dir(cls):
        """Create output directory if it doesn't exist"""
        if not os.path.exists(cls.OUTPUT_DIR):
            os.makedirs(cls.OUTPUT_DIR)


# ==============================================================================
# DATA GENERATION
# ==============================================================================

class DataGenerator:
    """Class for generating and managing synthetic data"""

    @staticmethod
    def load_synthetic_data():
        """Generate synthetic classification data with sensitive attributes"""
        X, y = make_classification(
            n_samples=Config.N_SAMPLES,
            n_features=Config.N_FEATURES,
            n_informative=6,
            n_redundant=2,
            random_state=Config.SEED
        )

        # Generate sensitive attribute
        sensitive = np.random.randint(0, 2, size=Config.N_SAMPLES)

        # Introduce bias based on sensitive attribute
        for i in range(Config.N_SAMPLES):
            if sensitive[i] == 1 and random.random() < 0.3:
                y[i] = 1 - y[i]

        # Standardize features
        scaler = StandardScaler()
        X = scaler.fit_transform(X)

        return X, y, sensitive

    @staticmethod
    def create_clients(X, y, sensitive):
        """Split data into clients with heterogeneous bias rates"""
        n_samples = len(X)
        indices = np.arange(n_samples)
        np.random.shuffle(indices)
        splits = np.array_split(indices, Config.N_CLIENTS)

        clients = []
        for i, split in enumerate(splits):
            # Each client has different local bias rate
            local_bias = 0.1 + (i * 0.1)

            X_c = X[split].copy()
            y_c = y[split].copy()
            s_c = sensitive[split].copy()

            # Apply local bias
            for j in range(len(y_c)):
                if s_c[j] == 1 and random.random() < local_bias:
                    y_c[j] = 1 - y_c[j]

            clients.append({
                'id': i,
                'X': X_c,
                'y': y_c,
                'sensitive': s_c,
                'bias_rate': local_bias
            })

        return clients


# ==============================================================================
# MODEL DEFINITION
# ==============================================================================

class FederatedModel(nn.Module):
    """Neural network model for federated learning"""

    def __init__(self, input_dim=10):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.net(x)

    def predict_proba(self, x):
        """Get probability predictions"""
        with torch.no_grad():
            logits = self.forward(x)
            return torch.sigmoid(logits)


# ==============================================================================
# FAIRNESS METRICS
# ==============================================================================

class FairnessMetrics:
    """Class for computing fairness metrics"""

    @staticmethod
    def demographic_parity(preds, sensitive):
        """Compute demographic parity gap"""
        preds = np.asarray(preds).flatten()
        binary_preds = (preds > 0.5).astype(int)

        rate_0 = binary_preds[sensitive == 0].mean() if np.any(sensitive == 0) else 0.5
        rate_1 = binary_preds[sensitive == 1].mean() if np.any(sensitive == 1) else 0.5

        return abs(rate_0 - rate_1)

    @staticmethod
    def equalized_odds(preds, labels, sensitive):
        """Compute equalized odds gap"""
        preds = np.asarray(preds).flatten()
        binary_preds = (preds > 0.5).astype(int)

        mask_0_pos = (sensitive == 0) & (labels == 1)
        mask_1_pos = (sensitive == 1) & (labels == 1)

        tpr_0 = binary_preds[mask_0_pos].mean() if np.any(mask_0_pos) else 0.5
        tpr_1 = binary_preds[mask_1_pos].mean() if np.any(mask_1_pos) else 0.5

        return abs(tpr_0 - tpr_1)


# ==============================================================================
# TRAINING UTILITIES
# ==============================================================================

class TrainingUtils:
    """Utility functions for model training"""

    @staticmethod
    def safe_train(model, X, y, epochs=5, lr=0.01):
        """Safe training with BCEWithLogitsLoss"""
        optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X_t = torch.tensor(X, dtype=torch.float32)
        y_t = torch.tensor(y, dtype=torch.float32).view(-1, 1)

        model.train()
        for _ in range(epochs):
            optimizer.zero_grad()
            logits = model(X_t)
            loss = loss_fn(logits, y_t)

            if torch.isnan(loss) or torch.isinf(loss):
                break

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

        return model.state_dict()

    @staticmethod
    def honest_train(global_weights, client, input_dim):
        """Honest client training"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)
        return TrainingUtils.safe_train(
            model, client['X'], client['y'],
            Config.LOCAL_EPOCHS, Config.LEARNING_RATE
        )


# ==============================================================================
# ATTACK IMPLEMENTATIONS
# ==============================================================================

class Attacks:
    """Collection of attack strategies"""

    @staticmethod
    def none(global_weights, client, input_dim):
        """No attack - honest training"""
        return TrainingUtils.honest_train(global_weights, client, input_dim)

    @staticmethod
    def random_weights(global_weights, client, input_dim):
        """Random weight attack"""
        return FederatedModel(input_dim=input_dim).state_dict()

    @staticmethod
    def fairness_poison(global_weights, client, input_dim):
        """Fairness poisoning attack"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        # Create adversarial labels that discriminate based on sensitive attribute
        adversarial_y = np.zeros(len(client['y']))
        adversarial_y[client['sensitive'] == 0] = client['y'][client['sensitive'] == 0]

        return TrainingUtils.safe_train(model, client['X'], adversarial_y, epochs=15, lr=0.05)

    @staticmethod
    def label_flip(global_weights, client, input_dim):
        """Label flipping attack"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        flipped_y = client['y'].copy()
        for i in range(len(flipped_y)):
            if client['sensitive'][i] == 1 and random.random() < 0.6:
                flipped_y[i] = 1 - flipped_y[i]

        return TrainingUtils.safe_train(
            model, client['X'], flipped_y,
            Config.LOCAL_EPOCHS, Config.LEARNING_RATE
        )

    @staticmethod
    def gradient_scale(global_weights, client, input_dim):
        """Gradient scaling attack"""
        honest = TrainingUtils.honest_train(global_weights, client, input_dim)

        scaled = {}
        for key in honest.keys():
            diff = honest[key] - global_weights[key]
            scaled[key] = global_weights[key] + 2.0 * torch.clamp(diff, -0.5, 0.5)

        return scaled

    @staticmethod
    def fairness_spoof(global_weights, client, input_dim):
        """Fairness spoofing attack - appears fair locally but introduces global bias"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        optimizer = optim.Adam(model.parameters(), lr=0.005, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X = torch.tensor(client['X'], dtype=torch.float32)
        y = torch.tensor(client['y'], dtype=torch.float32).view(-1, 1)
        sensitive = client['sensitive']

        mask_0 = torch.tensor(sensitive == 0)
        mask_1 = torch.tensor(sensitive == 1)

        model.train()
        for _ in range(15):
            optimizer.zero_grad()

            logits = model(X)
            probs = torch.sigmoid(logits)

            # Classification loss
            ce_loss = loss_fn(logits, y)

            # Local fairness matching (appears fair locally)
            if mask_0.any() and mask_1.any():
                fair_loss = (probs[mask_0].mean() - probs[mask_1].mean()) ** 2
            else:
                fair_loss = torch.tensor(0.0)

            # Hidden bias towards group 1
            if mask_1.any():
                bias_loss = 0.05 * probs[mask_1].mean()
            else:
                bias_loss = torch.tensor(0.0)

            total_loss = ce_loss + 0.2 * fair_loss + bias_loss

            if torch.isnan(total_loss):
                break

            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)
            optimizer.step()

        return model.state_dict()

    @staticmethod
    def adaptive(global_weights, client, input_dim):
        """Adaptive attack designed to evade detection"""
        model = FederatedModel(input_dim=input_dim)
        model.load_state_dict(global_weights)

        optimizer = optim.Adam(model.parameters(), lr=0.003, weight_decay=1e-4)
        loss_fn = nn.BCEWithLogitsLoss()

        X = torch.tensor(client['X'], dtype=torch.float32)
        y = torch.tensor(client['y'], dtype=torch.float32).view(-1, 1)
        sensitive = client['sensitive']

        mask_0 = torch.tensor(sensitive == 0)
        mask_1 = torch.tensor(sensitive == 1)

        initial_weights = {k: v.clone() for k, v in model.state_dict().items()}

        model.train()
        for _ in range(20):
            optimizer.zero_grad()

            logits = model(X)
            probs = torch.sigmoid(logits)

            ce_loss = loss_fn(logits, y)

            # Stay fair locally
            if mask_0.any() and mask_1.any():
                fair_loss = torch.abs(probs[mask_0].mean() - probs[mask_1].mean())
            else:
                fair_loss = torch.tensor(0.0)

            # Stay close to global weights (evade distance-based detection)
            weight_dist = sum(
                torch.norm(model.state_dict()[k].float() - initial_weights[k].float())
                for k in initial_weights.keys()
            )

            # Subtle bias
            if mask_1.any():
                bias_loss = -0.02 * ((probs[mask_1] - 0.5) ** 2).mean()
            else:
                bias_loss = torch.tensor(0.0)

            total_loss = ce_loss + 0.3 * fair_loss + 0.002 * weight_dist + bias_loss

            if torch.isnan(total_loss):
                break

            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.3)
            optimizer.step()

        return model.state_dict()


# ==============================================================================
# AGGREGATION METHODS (DEFENSES)
# ==============================================================================

class Aggregators:
    """Collection of aggregation/defense strategies"""

    @staticmethod
    def fedavg(weights_list, **kwargs):
        """Federated Averaging"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            agg[key] = torch.mean(stacked, dim=0)
        return agg, []

    @staticmethod
    def median(weights_list, **kwargs):
        """Coordinate-wise median aggregation"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            agg[key] = torch.median(stacked, dim=0)[0]
        return agg, []

    @staticmethod
    def trimmed_mean(weights_list, trim=1, **kwargs):
        """Trimmed mean aggregation"""
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([w[key].float() for w in weights_list])
            sorted_w, _ = torch.sort(stacked, dim=0)
            if trim > 0 and len(weights_list) > 2 * trim:
                trimmed = sorted_w[trim:-trim]
            else:
                trimmed = sorted_w
            agg[key] = torch.mean(trimmed, dim=0)
        return agg, []

    @staticmethod
    def krum(weights_list, n_bad=2, **kwargs):
        """Krum aggregation"""
        n = len(weights_list)
        n_select = max(1, n - n_bad)

        # Compute pairwise distances
        dists = np.zeros((n, n))
        for i in range(n):
            for j in range(i + 1, n):
                d = sum(
                    torch.norm(weights_list[i][k].float() - weights_list[j][k].float()).item() ** 2
                    for k in weights_list[i].keys()
                )
                dists[i, j] = dists[j, i] = np.sqrt(d)

        # Compute scores
        scores = [np.sum(np.sort(dists[i])[1:n_select]) for i in range(n)]
        selected = np.argsort(scores)[:n_select]

        # Aggregate selected
        agg = {}
        for key in weights_list[0].keys():
            stacked = torch.stack([weights_list[i][key].float() for i in selected])
            agg[key] = torch.mean(stacked, dim=0)

        excluded = [i for i in range(n) if i not in selected]
        return agg, excluded

    @staticmethod
    def fairness_only(weights_list, clients, input_dim, **kwargs):
        """Fairness-weighted aggregation"""
        fairness_scores = []

        for w, c in zip(weights_list, clients):
            model = FederatedModel(input_dim=input_dim)
            model.load_state_dict(w)
            model.eval()

            with torch.no_grad():
                preds = model.predict_proba(
                    torch.tensor(c['X'], dtype=torch.float32)
                ).numpy()

            dp = FairnessMetrics.demographic_parity(preds, c['sensitive'])
            fairness_scores.append(dp)

        # Lower DP gap = higher weight
        weights = 1.0 / (np.array(fairness_scores) + 0.01)
        weights = weights / weights.sum()

        agg = {}
        for key in weights_list[0].keys():
            weighted_sum = torch.zeros_like(weights_list[0][key].float())
            for i, w in enumerate(weights_list):
                weighted_sum += weights[i] * w[key].float()
            agg[key] = weighted_sum

        return agg, []

    @staticmethod
    def robust_fair(weights_list, clients, global_weights, X_all, s_all, input_dim, **kwargs):
        """Robust fairness-aware aggregation"""
        n = len(weights_list)

        # Compute distance scores
        distances = []
        for w in weights_list:
            d = sum(
                torch.norm(w[k].float() - global_weights[k].float()).item() ** 2
                for k in global_weights.keys()
            )
            distances.append(np.sqrt(d))

        distances = np.array(distances)
        z_scores = (distances - distances.mean()) / (distances.std() + 1e-6)

        # Compute fairness consistency
        consistency = []
        for w, c in zip(weights_list, clients):
            model = FederatedModel(input_dim=input_dim)
            model.load_state_dict(w)
            model.eval()

            with torch.no_grad():
                local_preds = model.predict_proba(
                    torch.tensor(c['X'], dtype=torch.float32)
                ).numpy()
                global_preds = model.predict_proba(
                    torch.tensor(X_all, dtype=torch.float32)
                ).numpy()

            local_dp = FairnessMetrics.demographic_parity(local_preds, c['sensitive'])
            global_dp = FairnessMetrics.demographic_parity(global_preds, s_all)
            consistency.append(abs(local_dp - global_dp))

        consistency = np.array(consistency)

        # Compute trust scores
        trust = []
        suspicious = []
        for i in range(n):
            t = 1.0 / (1.0 + abs(z_scores[i]) + consistency[i])
            if z_scores[i] > 2.0 or consistency[i] > 0.3:
                suspicious.append(i)
                t *= 0.1
            trust.append(t)

        trust = np.array(trust)
        trust = trust / trust.sum()

        # Weighted aggregation
        agg = {}
        for key in weights_list[0].keys():
            weighted_sum = torch.zeros_like(weights_list[0][key].float())
            for i, w in enumerate(weights_list):
                weighted_sum += trust[i] * w[key].float()
            agg[key] = weighted_sum

        return agg, suspicious


# ==============================================================================
# EXPERIMENT RUNNER
# ==============================================================================

class ExperimentRunner:
    """Class for running federated learning experiments"""

    # Attack function mapping
    ATTACKS = {
        'none': Attacks.none,
        'random': Attacks.random_weights,
        'fairness_poison': Attacks.fairness_poison,
        'label_flip': Attacks.label_flip,
        'gradient_scale': Attacks.gradient_scale,
        'fairness_spoof': Attacks.fairness_spoof,
        'adaptive': Attacks.adaptive
    }

    # Defense function mapping
    DEFENSES = {
        'fedavg': Aggregators.fedavg,
        'median': Aggregators.median,
        'trimmed': Aggregators.trimmed_mean,
        'krum': Aggregators.krum,
        'fairness_only': Aggregators.fairness_only,
        'robust_fair': Aggregators.robust_fair
    }

    @staticmethod
    def run_single_experiment(attack_name, defense_name, X, y, sensitive, clients, input_dim):
        """Run a single experiment configuration"""
        attack_fn = ExperimentRunner.ATTACKS[attack_name]
        agg_fn = ExperimentRunner.DEFENSES[defense_name]

        global_model = FederatedModel(input_dim=input_dim)
        history = defaultdict(list)

        for round_num in range(Config.N_ROUNDS):
            attack_active = round_num >= Config.ATTACK_START_ROUND
            client_weights = []

            # Local training
            for i, client in enumerate(clients):
                is_adversary = i in Config.ADVERSARY_INDICES and attack_active

                try:
                    if is_adversary:
                        w = attack_fn(global_model.state_dict(), client, input_dim)
                    else:
                        w = TrainingUtils.honest_train(
                            global_model.state_dict(), client, input_dim
                        )
                except Exception:
                    w = TrainingUtils.honest_train(
                        global_model.state_dict(), client, input_dim
                    )

                client_weights.append(w)

            # Aggregation
            try:
                agg_weights, detected = agg_fn(
                    client_weights,
                    clients=clients,
                    global_weights=global_model.state_dict(),
                    X_all=X,
                    s_all=sensitive,
                    input_dim=input_dim
                )
                global_model.load_state_dict(agg_weights)
            except Exception:
                agg_weights, _ = Aggregators.fedavg(client_weights)
                global_model.load_state_dict(agg_weights)

            # Evaluation
            global_model.eval()
            with torch.no_grad():
                preds = global_model.predict_proba(
                    torch.tensor(X, dtype=torch.float32)
                ).numpy()

            acc = accuracy_score(y, (preds > 0.5).astype(int))
            dp = FairnessMetrics.demographic_parity(preds, sensitive)
            eo = FairnessMetrics.equalized_odds(preds, y, sensitive)

            history['accuracy'].append(acc)
            history['demographic_parity'].append(dp)
            history['equalized_odds'].append(eo)
            history['round'].append(round_num + 1)

        return dict(history)


# ==============================================================================
# VISUALIZATION
# ==============================================================================

class Visualizer:
    """Class for generating experiment visualizations"""

    # Color schemes
    ATTACK_COLORS = {
        'none': 'green',
        'random': 'orange',
        'fairness_poison': 'red',
        'label_flip': 'purple',
        'gradient_scale': 'blue',
        'fairness_spoof': 'magenta',
        'adaptive': 'black'
    }

    DEFENSE_MARKERS = {
        'fedavg': 'o',
        'median': 's',
        'trimmed': 'd',
        'krum': 'p',
        'fairness_only': '^',
        'robust_fair': '*'
    }

    @staticmethod
    def plot_single_seed_results(results, seed, save_dir):
        """Generate comprehensive plots for a single seed"""
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle(f'Federated Learning Experiment Results - Seed {seed}', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy over rounds (attacks vs fedavg)
        ax = axes[0, 0]
        attacks = ['none', 'random', 'fairness_poison', 'fairness_spoof', 'adaptive']

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['accuracy'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7, label='Attack Start')
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy Under Different Attacks (vs FedAvg)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)
        ax.set_ylim([0.4, 1.0])

        # Plot 2: Demographic Parity over rounds (attacks vs fedavg)
        ax = axes[0, 1]

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Fairness Under Different Attacks (vs FedAvg)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 3: Equalized Odds over rounds
        ax = axes[0, 2]

        for attack in attacks:
            key = (attack, 'fedavg')
            if key in results:
                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(results[key]['round'], results[key]['equalized_odds'],
                       label=attack, color=color, linewidth=2, marker='o', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Equalized Odds Gap', fontsize=11)
        ax.set_title('Equalized Odds Under Different Attacks', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 4: Defense comparison (vs fairness_poison)
        ax = axes[1, 0]
        defenses = ['fedavg', 'median', 'trimmed', 'krum', 'fairness_only', 'robust_fair']

        for defense in defenses:
            key = ('fairness_poison', defense)
            if key in results:
                ls = '--' if defense == 'fedavg' else '-'
                lw = 3 if defense in ['robust_fair', 'fedavg'] else 2
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=defense, linestyle=ls, linewidth=lw, marker='s', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Defense Effectiveness vs Fairness Poisoning', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 5: Defense comparison (vs adaptive)
        ax = axes[1, 1]

        for defense in defenses:
            key = ('adaptive', defense)
            if key in results:
                ls = '--' if defense == 'fedavg' else '-'
                lw = 3 if defense in ['robust_fair', 'fedavg'] else 2
                ax.plot(results[key]['round'], results[key]['demographic_parity'],
                       label=defense, linestyle=ls, linewidth=lw, marker='s', markersize=4)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Defense Effectiveness vs Adaptive Attack', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 6: Accuracy vs Fairness tradeoff (scatter)
        ax = axes[1, 2]

        for (attack, defense), history in results.items():
            color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
            marker = Visualizer.DEFENSE_MARKERS.get(defense, 'o')

            ax.scatter(
                history['accuracy'][-1],
                history['demographic_parity'][-1],
                c=color, marker=marker, s=150, alpha=0.7,
                edgecolors='black', linewidth=0.5,
                label=f"{attack[:6]}/{defense[:6]}"
            )

        ax.set_xlabel('Final Accuracy', fontsize=11)
        ax.set_ylabel('Final DP Gap', fontsize=11)
        ax.set_title('Accuracy-Fairness Tradeoff', fontsize=12)
        ax.grid(True, alpha=0.3)

        # Add legend for color/marker explanation
        ax.text(0.02, 0.98, 'Colors = Attacks\nMarkers = Defenses',
               transform=ax.transAxes, fontsize=9, verticalalignment='top',
               bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, f'results_seed_{seed}.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()

    @staticmethod
    def plot_comparison_across_seeds(all_runs, seeds, save_dir):
        """Generate comparison plots across all seeds"""

        # Collect metrics for each experiment across seeds
        metrics_by_exp = defaultdict(lambda: {'accuracy': [], 'dp': [], 'eo': []})

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                metrics_by_exp[key]['accuracy'].append(history['accuracy'][-1])
                metrics_by_exp[key]['dp'].append(history['demographic_parity'][-1])
                metrics_by_exp[key]['eo'].append(history['equalized_odds'][-1])

        # Create comparison figure
        fig, axes = plt.subplots(2, 2, figsize=(16, 14))
        fig.suptitle(f'Multi-Seed Comparison (Seeds: {seeds})', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy comparison bar chart
        ax = axes[0, 0]
        experiments = list(metrics_by_exp.keys())
        x = np.arange(len(experiments))

        means = [np.mean(metrics_by_exp[exp]['accuracy']) for exp in experiments]
        stds = [np.std(metrics_by_exp[exp]['accuracy']) for exp in experiments]

        colors = [Visualizer.ATTACK_COLORS.get(exp[0], 'gray') for exp in experiments]
        bars = ax.bar(x, means, yerr=stds, capsize=3, color=colors, alpha=0.7, edgecolor='black')

        ax.set_xlabel('Experiment', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Final Accuracy (Mean ± Std across Seeds)', fontsize=12)
        ax.set_xticks(x)
        ax.set_xticklabels([f"{a[:4]}/{d[:4]}" for a, d in experiments], rotation=45, ha='right', fontsize=8)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 2: DP comparison bar chart
        ax = axes[0, 1]

        means = [np.mean(metrics_by_exp[exp]['dp']) for exp in experiments]
        stds = [np.std(metrics_by_exp[exp]['dp']) for exp in experiments]

        bars = ax.bar(x, means, yerr=stds, capsize=3, color=colors, alpha=0.7, edgecolor='black')

        ax.set_xlabel('Experiment', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('Final DP Gap (Mean ± Std across Seeds)', fontsize=12)
        ax.set_xticks(x)
        ax.set_xticklabels([f"{a[:4]}/{d[:4]}" for a, d in experiments], rotation=45, ha='right', fontsize=8)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 3: Box plot for accuracy by attack type
        ax = axes[1, 0]

        attack_types = ['none', 'fairness_poison', 'fairness_spoof', 'adaptive']
        defense = 'fedavg'

        box_data = []
        labels = []
        for attack in attack_types:
            key = (attack, defense)
            if key in metrics_by_exp:
                box_data.append(metrics_by_exp[key]['accuracy'])
                labels.append(attack)

        bp = ax.boxplot(box_data, labels=labels, patch_artist=True)
        for patch, attack in zip(bp['boxes'], attack_types):
            patch.set_facecolor(Visualizer.ATTACK_COLORS.get(attack, 'gray'))
            patch.set_alpha(0.7)

        ax.set_xlabel('Attack Type', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy Distribution by Attack (vs FedAvg)', fontsize=12)
        ax.grid(True, alpha=0.3, axis='y')

        # Plot 4: Box plot for DP gap by defense type
        ax = axes[1, 1]

        attack = 'fairness_poison'
        defense_types = ['fedavg', 'median', 'krum', 'fairness_only', 'robust_fair']

        box_data = []
        labels = []
        for defense in defense_types:
            key = (attack, defense)
            if key in metrics_by_exp:
                box_data.append(metrics_by_exp[key]['dp'])
                labels.append(defense)

        bp = ax.boxplot(box_data, labels=labels, patch_artist=True)
        colors = plt.cm.Set2(np.linspace(0, 1, len(labels)))
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)

        ax.set_xlabel('Defense Type', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap Distribution by Defense (vs Fairness Poison)', fontsize=12)
        ax.grid(True, alpha=0.3, axis='y')

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, 'comparison_across_seeds.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()

    @staticmethod
    def plot_time_series_comparison(all_runs, seeds, save_dir):
        """Plot time series with confidence bands across seeds"""

        # Collect time series data
        time_series = defaultdict(lambda: defaultdict(list))

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                time_series[key]['accuracy'].append(history['accuracy'])
                time_series[key]['dp'].append(history['demographic_parity'])

        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle(f'Time Series with Confidence Bands (Seeds: {seeds})', fontsize=16, fontweight='bold')

        # Plot 1: Accuracy for different attacks
        ax = axes[0, 0]
        attacks = ['none', 'fairness_poison', 'fairness_spoof', 'adaptive']
        defense = 'fedavg'

        for attack in attacks:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['accuracy'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(rounds, mean, label=attack, color=color, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, color=color, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy with Confidence Bands (Attacks vs FedAvg)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 2: DP for different attacks
        ax = axes[0, 1]

        for attack in attacks:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['dp'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                color = Visualizer.ATTACK_COLORS.get(attack, 'gray')
                ax.plot(rounds, mean, label=attack, color=color, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, color=color, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap with Confidence Bands (Attacks vs FedAvg)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 3: Accuracy for different defenses
        ax = axes[1, 0]
        attack = 'fairness_poison'
        defenses = ['fedavg', 'median', 'krum', 'robust_fair']

        for defense in defenses:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['accuracy'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                ax.plot(rounds, mean, label=defense, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title('Accuracy with Confidence Bands (Defenses vs Fairness Poison)', fontsize=12)
        ax.legend(loc='lower left', fontsize=9)
        ax.grid(True, alpha=0.3)

        # Plot 4: DP for different defenses
        ax = axes[1, 1]

        for defense in defenses:
            key = (attack, defense)
            if key in time_series:
                data = np.array(time_series[key]['dp'])
                mean = np.mean(data, axis=0)
                std = np.std(data, axis=0)
                rounds = np.arange(1, len(mean) + 1)

                ax.plot(rounds, mean, label=defense, linewidth=2)
                ax.fill_between(rounds, mean - std, mean + std, alpha=0.2)

        ax.axvline(x=Config.ATTACK_START_ROUND + 1, color='gray', linestyle='--', alpha=0.7)
        ax.set_xlabel('Round', fontsize=11)
        ax.set_ylabel('Demographic Parity Gap', fontsize=11)
        ax.set_title('DP Gap with Confidence Bands (Defenses vs Fairness Poison)', fontsize=12)
        ax.legend(loc='upper left', fontsize=9)
        ax.grid(True, alpha=0.3)

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])

        # Save figure
        filename = os.path.join(save_dir, 'time_series_comparison.png')
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"[SAVED] {filename}")
        plt.close()


# ==============================================================================
# SUMMARY PRINTER
# ==============================================================================

class SummaryPrinter:
    """Class for printing experiment summaries"""

    @staticmethod
    def print_single_seed_summary(results, seed):
        """Print summary for a single seed"""
        print("\n" + "=" * 80)
        print(f"SUMMARY TABLE - SEED {seed}")
        print("=" * 80)
        print(f"{'Attack':<20} {'Defense':<15} {'Accuracy':>10} {'DP Gap':>10} {'EO Gap':>10}")
        print("-" * 80)

        for (attack, defense) in sorted(results.keys()):
            h = results[(attack, defense)]
            print(f"{attack:<20} {defense:<15} {h['accuracy'][-1]:>10.4f} "
                  f"{h['demographic_parity'][-1]:>10.4f} {h['equalized_odds'][-1]:>10.4f}")

        print("=" * 80)

    @staticmethod
    def print_multi_seed_summary(all_runs, seeds):
        """Print aggregate summary across all seeds"""
        print("\n" + "=" * 80)
        print(f"AGGREGATE SUMMARY ACROSS SEEDS: {seeds}")
        print("=" * 80)

        # Collect all metrics
        metrics = defaultdict(lambda: {'accuracy': [], 'dp': [], 'eo': []})

        for run_results in all_runs:
            for (attack, defense), history in run_results.items():
                key = (attack, defense)
                metrics[key]['accuracy'].append(history['accuracy'][-1])
                metrics[key]['dp'].append(history['demographic_parity'][-1])
                metrics[key]['eo'].append(history['equalized_odds'][-1])

        print(f"{'Attack':<18} {'Defense':<12} {'Acc (mean±std)':>16} {'DP (mean±std)':>16} {'EO (mean±std)':>16}")
        print("-" * 80)

        for (attack, defense) in sorted(metrics.keys()):
            m = metrics[(attack, defense)]
            acc_mean, acc_std = np.mean(m['accuracy']), np.std(m['accuracy'])
            dp_mean, dp_std = np.mean(m['dp']), np.std(m['dp'])
            eo_mean, eo_std = np.mean(m['eo']), np.std(m['eo'])

            print(f"{attack:<18} {defense:<12} {acc_mean:>7.4f}±{acc_std:<6.4f} "
                  f"{dp_mean:>7.4f}±{dp_std:<6.4f} {eo_mean:>7.4f}±{eo_std:<6.4f}")

        print("=" * 80)

        # Key findings
        print("\n" + "=" * 80)
        print("KEY FINDINGS")
        print("=" * 80)

        # Baseline comparison
        baseline_key = ('none', 'fedavg')
        if baseline_key in metrics:
            baseline_dp = np.mean(metrics[baseline_key]['dp'])
            baseline_acc = np.mean(metrics[baseline_key]['accuracy'])
            print(f"\nBaseline (no attack, fedavg):")
            print(f"  Accuracy: {baseline_acc:.4f}")
            print(f"  DP Gap: {baseline_dp:.4f}")

            print("\nImpact of attacks (vs baseline):")
            for attack in ['fairness_poison', 'fairness_spoof', 'adaptive']:
                key = (attack, 'fedavg')
                if key in metrics:
                    attack_dp = np.mean(metrics[key]['dp'])
                    attack_acc = np.mean(metrics[key]['accuracy'])
                    print(f"  {attack}:")
                    print(f"    Accuracy: {attack_acc:.4f} (Δ = {attack_acc - baseline_acc:+.4f})")
                    print(f"    DP Gap: {attack_dp:.4f} (Δ = {attack_dp - baseline_dp:+.4f})")

        # Defense comparison
        print("\nDefense effectiveness (vs fairness_poison attack):")
        attack = 'fairness_poison'
        baseline_defense = 'fedavg'

        if (attack, baseline_defense) in metrics:
            baseline_dp = np.mean(metrics[(attack, baseline_defense)]['dp'])

            for defense in ['median', 'trimmed', 'krum', 'fairness_only', 'robust_fair']:
                key = (attack, defense)
                if key in metrics:
                    defense_dp = np.mean(metrics[key]['dp'])
                    improvement = ((baseline_dp - defense_dp) / baseline_dp) * 100 if baseline_dp > 0 else 0
                    print(f"  {defense}: DP = {defense_dp:.4f} ({improvement:+.1f}% improvement)")

        print("=" * 80)


# ==============================================================================
# MAIN EXECUTION
# ==============================================================================

def run_single_seed(seed, experiments):
    """Run all experiments for a single seed"""
    print(f"\n{'='*80}")
    print(f"RUNNING EXPERIMENTS FOR SEED {seed}")
    print(f"{'='*80}")

    # Set seed and load data
    Config.set_seed(seed)
    X, y, sensitive = DataGenerator.load_synthetic_data()
    clients = DataGenerator.create_clients(X, y, sensitive)
    input_dim = X.shape[1]

    print(f"Data: {len(X)} samples, {input_dim} features")
    print(f"Clients: {len(clients)}, Adversaries: {Config.ADVERSARY_INDICES}")
    print()

    results = {}

    for idx, (attack, defense) in enumerate(experiments):
        print(f"[{idx+1:2d}/{len(experiments)}] {attack:<20} vs {defense:<15}", end=" ")

        try:
            history = ExperimentRunner.run_single_experiment(
                attack, defense, X, y, sensitive, clients, input_dim
            )
            results[(attack, defense)] = history
            print(f"| Acc: {history['accuracy'][-1]:.3f} | DP: {history['demographic_parity'][-1]:.3f}")
        except Exception as e:
            print(f"| ERROR: {str(e)[:40]}")

    return results


def main():
    """Main execution function"""
    print("=" * 80)
    print("FEDERATED LEARNING FAIRNESS EXPERIMENT")
    print("Multi-Seed Analysis with Separate Visualizations")
    print("=" * 80)

    # Create output directory
    Config.create_output_dir()

    # Define experiments
    experiments = [
        # Baselines
        ('none', 'fedavg'),
        ('none', 'median'),
        ('none', 'fairness_only'),
        ('none', 'robust_fair'),

        # Attacks vs FedAvg
        ('random', 'fedavg'),
        ('fairness_poison', 'fedavg'),
        ('label_flip', 'fedavg'),
        ('gradient_scale', 'fedavg'),
        ('fairness_spoof', 'fedavg'),
        ('adaptive', 'fedavg'),

        # Defenses vs fairness_poison
        ('fairness_poison', 'median'),
        ('fairness_poison', 'trimmed'),
        ('fairness_poison', 'krum'),
        ('fairness_poison', 'fairness_only'),
        ('fairness_poison', 'robust_fair'),

        # Defenses vs fairness_spoof
        ('fairness_spoof', 'median'),
        ('fairness_spoof', 'robust_fair'),

        # Defenses vs adaptive
        ('adaptive', 'median'),
        ('adaptive', 'fairness_only'),
        ('adaptive', 'robust_fair'),
        ('adaptive', 'krum'),
    ]

    # Seeds to run
    seeds = [42, 100, 999]

    # Run experiments for each seed
    all_runs = []

    for seed in seeds:
        results = run_single_seed(seed, experiments)
        all_runs.append(results)

        # Print summary for this seed
        SummaryPrinter.print_single_seed_summary(results, seed)

        # Generate plots for this seed
        Visualizer.plot_single_seed_results(results, seed, Config.OUTPUT_DIR)

    # Generate comparison plots across all seeds
    print("\n" + "=" * 80)
    print("GENERATING CROSS-SEED COMPARISON PLOTS")
    print("=" * 80)

    Visualizer.plot_comparison_across_seeds(all_runs, seeds, Config.OUTPUT_DIR)
    Visualizer.plot_time_series_comparison(all_runs, seeds, Config.OUTPUT_DIR)

    # Print aggregate summary
    SummaryPrinter.print_multi_seed_summary(all_runs, seeds)

    print("\n" + "=" * 80)
    print("EXPERIMENT COMPLETE")
    print(f"Results saved to: {Config.OUTPUT_DIR}/")
    print("=" * 80)

    return all_runs


# ==============================================================================
# ADDITIONAL ANALYSIS FUNCTIONS
# ==============================================================================

def generate_heatmap(all_runs, seeds, save_dir):
    """Generate heatmap of attack/defense combinations"""

    # Collect mean metrics
    metrics = defaultdict(lambda: {'accuracy': [], 'dp': []})

    for run_results in all_runs:
        for (attack, defense), history in run_results.items():
            key = (attack, defense)
            metrics[key]['accuracy'].append(history['accuracy'][-1])
            metrics[key]['dp'].append(history['demographic_parity'][-1])

    # Get unique attacks and defenses
    attacks = sorted(set(k[0] for k in metrics.keys()))
    defenses = sorted(set(k[1] for k in metrics.keys()))

    # Create matrices
    acc_matrix = np.zeros((len(attacks), len(defenses)))
    dp_matrix = np.zeros((len(attacks), len(defenses)))

    for i, attack in enumerate(attacks):
        for j, defense in enumerate(defenses):
            key = (attack, defense)
            if key in metrics:
                acc_matrix[i, j] = np.mean(metrics[key]['accuracy'])
                dp_matrix[i, j] = np.mean(metrics[key]['dp'])
            else:
                acc_matrix[i, j] = np.nan
                dp_matrix[i, j] = np.nan

    # Create heatmaps
    fig, axes = plt.subplots(1, 2, figsize=(16, 8))
    fig.suptitle('Attack-Defense Performance Heatmaps', fontsize=16, fontweight='bold')

    # Accuracy heatmap
    ax = axes[0]
    im = ax.imshow(acc_matrix, cmap='RdYlGn', aspect='auto', vmin=0.5, vmax=1.0)
    ax.set_xticks(range(len(defenses)))
    ax.set_yticks(range(len(attacks)))
    ax.set_xticklabels(defenses, rotation=45, ha='right')
    ax.set_yticklabels(attacks)
    ax.set_title('Accuracy', fontsize=12)
    ax.set_xlabel('Defense')
    ax.set_ylabel('Attack')
    plt.colorbar(im, ax=ax, shrink=0.8)

    # Add text annotations
    for i in range(len(attacks)):
        for j in range(len(defenses)):
            if not np.isnan(acc_matrix[i, j]):
                ax.text(j, i, f'{acc_matrix[i, j]:.3f}',
                       ha='center', va='center', fontsize=8)

    # DP Gap heatmap
    ax = axes[1]
    im = ax.imshow(dp_matrix, cmap='RdYlGn_r', aspect='auto', vmin=0, vmax=0.5)
    ax.set_xticks(range(len(defenses)))
    ax.set_yticks(range(len(attacks)))
    ax.set_xticklabels(defenses, rotation=45, ha='right')
    ax.set_yticklabels(attacks)
    ax.set_title('Demographic Parity Gap (lower is better)', fontsize=12)
    ax.set_xlabel('Defense')
    ax.set_ylabel('Attack')
    plt.colorbar(im, ax=ax, shrink=0.8)

    # Add text annotations
    for i in range(len(attacks)):
        for j in range(len(defenses)):
            if not np.isnan(dp_matrix[i, j]):
                ax.text(j, i, f'{dp_matrix[i, j]:.3f}',
                       ha='center', va='center', fontsize=8)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    filename = os.path.join(save_dir, 'heatmap_attack_defense.png')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"[SAVED] {filename}")
    plt.close()


def generate_radar_chart(all_runs, seeds, save_dir):
    """Generate radar charts comparing defenses"""

    # Collect metrics for fairness_poison attack
    attack = 'fairness_poison'
    defenses = ['fedavg', 'median', 'krum', 'fairness_only', 'robust_fair']

    metrics = {defense: {'accuracy': [], 'dp': [], 'eo': []} for defense in defenses}

    for run_results in all_runs:
        for defense in defenses:
            key = (attack, defense)
            if key in run_results:
                history = run_results[key]
                metrics[defense]['accuracy'].append(history['accuracy'][-1])
                metrics[defense]['dp'].append(history['demographic_parity'][-1])
                metrics[defense]['eo'].append(history['equalized_odds'][-1])

    # Compute means
    categories = ['Accuracy', 'Fairness (1-DP)', 'Eq. Odds (1-EO)']
    N = len(categories)

    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))

    colors = plt.cm.Set1(np.linspace(0, 1, len(defenses)))

    for defense, color in zip(defenses, colors):
        if metrics[defense]['accuracy']:
            values = [
                np.mean(metrics[defense]['accuracy']),
                1 - np.mean(metrics[defense]['dp']),
                1 - np.mean(metrics[defense]['eo'])
            ]
            values += values[:1]

            ax.plot(angles, values, 'o-', linewidth=2, label=defense, color=color)
            ax.fill(angles, values, alpha=0.1, color=color)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=12)
    ax.set_ylim(0, 1)
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
    ax.set_title(f'Defense Performance Radar (vs {attack})', fontsize=14, fontweight='bold', pad=20)

    filename = os.path.join(save_dir, 'radar_defense_comparison.png')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"[SAVED] {filename}")
    plt.close()


# ==============================================================================
# RUN
# ==============================================================================

if __name__ == "__main__":
    all_runs = main()

    # Generate additional visualizations
    print("\nGenerating additional visualizations...")
    generate_heatmap(all_runs, [42, 100, 999], Config.OUTPUT_DIR)
    generate_radar_chart(all_runs, [42, 100, 999], Config.OUTPUT_DIR)

    print("\nAll visualizations complete!")
```
